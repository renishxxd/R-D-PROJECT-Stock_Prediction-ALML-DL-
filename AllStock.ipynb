```python
# NEW: Install necessary libraries in the Colab environment
!pip install yfinance transformers torch scikit-learn requests tqdm

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from datetime import datetime, timedelta
from tqdm import tqdm
import requests
import yfinance as yf
import warnings

warnings.filterwarnings('ignore')

# --- Constants ---
SEQ_LEN = 30
BATCH_SIZE = 32
EPOCHS = 30
LEARNING_RATE = 0.0001
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
# --- IMPORTANT: Replace with your own NewsAPI key ---
NEWS_API_KEY = '99d1937942dd430cb7e67bbd39c0a713' # <--- REPLACE THIS

print(f"Using device: {DEVICE}")

# --- Data Fetching & Cleaning ---
def fetch_stock_data(ticker="RELIANCE.NS"):
    """Fetches daily stock data for the last 5 years and cleans it."""
    df = yf.download(tickers=ticker, period="5y", interval="1d")

    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    # FIX: Corrected the syntax for the inplace argument.
    df.reset_index(inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    return df

def fetch_news(symbol="Reliance Industries"):
    """Fetches news from the last 30 days from NewsAPI."""
    if NEWS_API_KEY == 'YOUR_NEWS_API_KEY':
        print("Warning: NewsAPI key not provided. Skipping news fetching.")
        return pd.DataFrame(columns=["Date", "Headline"])

    to_date = datetime.today().strftime('%Y-%m-%d')
    from_date = (datetime.today() - timedelta(days=30)).strftime('%Y-%m-%d')
    url = f"https://newsapi.org/v2/everything?q={symbol}&from={from_date}&to={to_date}&language=en&sortBy=publishedAt&apiKey={NEWS_API_KEY}"
    try:
        res = requests.get(url).json()
        if res.get("status") == "ok":
            articles = res.get("articles", [])
            headlines = [{"Date": a['publishedAt'][:10], "Headline": a['title']} for a in articles]
            return pd.DataFrame(headlines)
        else:
            print(f"Error fetching news: {res.get('message')}")
            return pd.DataFrame(columns=["Date", "Headline"])
    except Exception as e:
        print(f"An exception occurred during news fetching: {e}")
        return pd.DataFrame(columns=["Date", "Headline"])

# MODIFIED: Function to fetch India-relevant events
def fetch_real_events(ticker_symbol, stock_df):
    """
    Fetches real financial and company-specific events relevant to the Indian market.
    - Earnings reports from yfinance
    - RBI MPC meeting dates (hard-coded for reliability)
    - Major Reliance events (if applicable)
    """
    all_events = []

    # 1. Fetch Earnings Dates (works for Indian stocks too)
    try:
        ticker_obj = yf.Ticker(ticker_symbol)
        earnings_dates = ticker_obj.get_earnings_dates(limit=20)
        if earnings_dates is not None and not earnings_dates.empty:
            earnings_df = pd.DataFrame({
                'Date': pd.to_datetime(earnings_dates.index.date),
                'Event_Type': 'Earnings'
            })
            all_events.append(earnings_df)
    except Exception as e:
        print(f"Could not fetch earnings dates: {e}")



    # 3. NEW: Add company-specific events for Reliance
    if ticker_symbol == 'RELIANCE.NS':
        reliance_event_dates = [
            '2020-07-15', # AGM 2020 (Jio Platforms announcements)
            '2021-06-24', # AGM 2021
            '2022-08-29', # AGM 2022 (5G rollout announcement)
            '2023-08-28', # AGM 2023
            '2024-08-29', # AGM 2024
            '2025-07-12'  # Hypothetical future date for AGM 2025
        ]
        reliance_events_df = pd.DataFrame({
            'Date': pd.to_datetime(reliance_event_dates),
            'Event_Type': 'Reliance AGM'
        })
        all_events.append(reliance_events_df)

    if not all_events:
        return pd.DataFrame(columns=['Date'])

    events_df = pd.concat(all_events, ignore_index=True)
    min_date, max_date = stock_df['Date'].min(), stock_df['Date'].max()
    events_df = events_df[(events_df['Date'] >= min_date) & (events_df['Date'] <= max_date)]
    return events_df.drop_duplicates(subset=['Date', 'Event_Type']).reset_index(drop=True)


# --- Load Data ---
# MODIFIED: Constants for Indian stock
TICKER = "RELIANCE.NS"
COMPANY_NAME = "Reliance Industries"
stock_df = fetch_stock_data(TICKER)
news_df = fetch_news(COMPANY_NAME)
events_df = fetch_real_events(TICKER, stock_df)


# --- FinBERT Sentiment Scoring ---
if not news_df.empty:
    try:
        tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
        model_fbert = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone").to(DEVICE)

        def get_sentiment(text):
            if not text or not isinstance(text, str): return 0.0
            inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True).to(DEVICE)
            with torch.no_grad():
                outputs = model_fbert(**inputs)
            scores = torch.nn.functional.softmax(outputs.logits, dim=1).cpu().numpy()[0]
            return scores[1] - scores[0]

        tqdm.pandas(desc="Calculating Sentiment")
        news_df['Sentiment'] = news_df['Headline'].progress_apply(get_sentiment)
        sentiment_df = news_df.groupby("Date")["Sentiment"].mean().reset_index()
        sentiment_df['Date'] = pd.to_datetime(sentiment_df['Date'])
    except Exception as e:
        print(f"Error loading FinBERT model or processing sentiment: {e}")
        sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])
else:
    sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])

# --- Feature Engineering ---
def add_features(df):
    df['Return'] = df['Close'].pct_change()
    df['MA10'] = df['Close'].rolling(window=10).mean()
    df['MA50'] = df['Close'].rolling(window=50).mean()
    delta = df['Close'].diff()
    up, down = delta.clip(lower=0), -1 * delta.clip(upper=0)
    ema_up, ema_down = up.ewm(com=13, adjust=False).mean(), down.ewm(com=13, adjust=False).mean()
    rs = ema_up / ema_down
    df['RSI'] = 100 - (100 / (1 + rs))
    df['MACD'] = df['Close'].ewm(span=12, adjust=False).mean() - df['Close'].ewm(span=26, adjust=False).mean()
    df['Volatility'] = df['Return'].rolling(window=21).std() * np.sqrt(252)
    return df

stock_df = add_features(stock_df)

# --- Merge Data ---
if not events_df.empty:
    event_dummies = pd.get_dummies(events_df['Event_Type'], prefix='Event', dtype=float)
    events_df = pd.concat([events_df['Date'], event_dummies], axis=1)
    events_df = events_df.groupby('Date').sum().reset_index()
    stock_df = pd.merge(stock_df, events_df, on='Date', how='left')

if not sentiment_df.empty:
    stock_df = pd.merge(stock_df, sentiment_df, on='Date', how='left')
else:
    stock_df['Sentiment'] = 0.0

event_cols = [col for col in stock_df.columns if 'Event_' in col]

# --- Preprocessing ---
stock_df['Target'] = stock_df['Return'].shift(-1)
stock_df['Sentiment'].fillna(0, inplace=True)
for col in event_cols:
    stock_df[col].fillna(0, inplace=True)

tech_cols = ['Return', 'MA10', 'MA50', 'RSI', 'MACD', 'Volatility']
stock_df[tech_cols] = stock_df[tech_cols].fillna(method='ffill').fillna(method='bfill')
stock_df.replace([np.inf, -np.inf], 0, inplace=True)
stock_df.dropna(inplace=True)

# --- Normalize & Split Data ---
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA10', 'MA50', 'RSI', 'MACD', 'Return', 'Volatility', 'Sentiment'] + event_cols

scaler = MinMaxScaler()
stock_df_scaled = stock_df.copy()
stock_df_scaled[features] = scaler.fit_transform(stock_df_scaled[features])

train_size = int(len(stock_df_scaled) * 0.85)
train_df = stock_df_scaled.iloc[:train_size].copy()
test_df = stock_df_scaled.iloc[train_size:].copy()
test_df_unscaled = stock_df.iloc[train_size:].copy()


# --- PyTorch Dataset ---
class StockDataset(Dataset):
    def __init__(self, df, df_unscaled, features, target_col, seq_len):
        self.X, self.y = [], []
        self.close_prices_for_eval = df_unscaled['Close'].values[seq_len-1:-1]

        feature_data = df[features].values
        target_data = df[target_col].values

        for i in range(seq_len, len(df)):
            self.X.append(feature_data[i-seq_len:i])
            self.y.append(target_data[i])

        self.X = np.array(self.X)
        self.y = np.array(self.y)

    def __len__(self): return len(self.X)
    def __getitem__(self, idx):
        return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)

train_dataset = StockDataset(train_df, train_df, features, 'Target', SEQ_LEN)
test_dataset = StockDataset(test_df, test_df_unscaled, features, 'Target', SEQ_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# --- Model Definitions ---
class LSTM_Attention(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(TLSTM_Attention, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.attention = nn.Linear(hidden_size, 1)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
        context = torch.sum(attn_weights * lstm_out, dim=1)
        return self.fc(context)

class GRU_Model(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(GRU_Model, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.gru(x)
        return self.fc(out[:, -1, :])

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.attn = nn.MultiheadAttention(embed_dim, num_heads, batch_first=True, dropout=rate)
        self.ffn = nn.Sequential(nn.Linear(embed_dim, ff_dim), nn.ReLU(), nn.Linear(ff_dim, embed_dim))
        self.layernorm1, self.layernorm2 = nn.LayerNorm(embed_dim, eps=1e-6), nn.LayerNorm(embed_dim, eps=1e-6)
        self.dropout1, self.dropout2 = nn.Dropout(rate), nn.Dropout(rate)
        self.fc_out = nn.Linear(embed_dim, 1)
    def forward(self, x):
        attn_output, _ = self.attn(x, x, x)
        out1 = self.layernorm1(x + self.dropout1(attn_output))
        ffn_output = self.ffn(out1)
        out2 = self.layernorm2(out1 + self.dropout2(ffn_output))
        return self.fc_out(out2.mean(dim=1))

# --- Training and Evaluation ---
def train_model(model, train_loader, epochs, lr):
    loss_fn = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.train()
    print(f"\n--- Training {model.__class__.__name__} ---")
    for epoch in range(epochs):
        total_loss = 0
        for X_batch, y_batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
            optimizer.zero_grad()
            preds = model(X_batch).squeeze(-1)
            loss = loss_fn(preds, y_batch)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch+1}/{epochs} - Loss: {total_loss / len(train_loader):.8f}")
    return model

def get_predictions(model, loader):
    model.eval()
    preds = []
    with torch.no_grad():
        for X_batch, _ in loader:
            preds.append(model(X_batch.to(DEVICE)).squeeze(-1).cpu().numpy())
    return np.concatenate(preds)

# --- Instantiate and Train All Models ---
input_size = len(features)
hidden_size = 64

# The number of features might change depending on the events found.
# Let's make the Transformer head count dynamic to avoid errors.
if input_size % 2 == 0:
    num_heads = 2
elif input_size % 3 == 0:
    num_heads = 3
else:
    num_heads = 1 # Fallback to 1 head if not divisible by 2 or 3

print(f"Total input features: {input_size}. Using {num_heads} attention heads for Transformer.")

tlstm_model = train_model(TLSTM_Attention(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
gru_model = train_model(GRU_Model(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
trans_model = train_model(TransformerBlock(input_size, num_heads, 64).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)

# --- Ensemble Prediction ---
lstm_preds = get_predictions(tlstm_model, test_loader)
gru_preds = get_predictions(gru_model, test_loader)
trans_preds = get_predictions(trans_model, test_loader)

final_predicted_returns = (tlstm_preds + gru_preds + trans_preds) / 3.0

# --- Convert Predicted Returns back to Prices for Evaluation ---
last_actual_prices = test_dataset.close_prices_for_eval
predicted_prices = last_actual_prices * (1 + final_predicted_returns)
actual_prices = test_df_unscaled['Close'].values[SEQ_LEN:]

# --- Evaluation ---
print("\n--- Final Ensemble Model Evaluation ---")
mae = mean_absolute_error(actual_prices, predicted_prices)
mse = mean_squared_error(actual_prices, predicted_prices)
r2 = r2_score(actual_prices, predicted_prices)

# MODIFIED: Changed currency to INR (₹)
print(f"Mean Absolute Error (MAE): ₹{mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R²): {r2:.4f}")

# --- Plot Results ---
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(15, 7))
plt.plot(test_df_unscaled['Date'][SEQ_LEN:], actual_prices, label='Actual Price', color='royalblue', linewidth=2)
plt.plot(test_df_unscaled['Date'][SEQ_LEN:], predicted_prices, label='Predicted Price', color='crimson', linestyle='--')
# MODIFIED: Title and Y-axis label updated for Indian stock and currency
plt.title(f"{TICKER} Stock Price Prediction: Actual vs. Predicted (Ensemble)", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Close Price (INR)", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# NEW: Install necessary libraries in the Colab environment
# !pip install yfinance transformers torch scikit-learn requests tqdm

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from datetime import datetime, timedelta
from tqdm import tqdm
import requests
import yfinance as yf
import warnings

warnings.filterwarnings('ignore')

# --- Constants ---
SEQ_LEN = 30
BATCH_SIZE = 32
EPOCHS = 30
LEARNING_RATE = 0.0001
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
# --- IMPORTANT: Replace with your own NewsAPI key ---
NEWS_API_KEY = '99d1937942dd430cb7e67bbd39c0a713' # <--- REPLACE THIS

print(f"Using device: {DEVICE}")

# --- Data Fetching & Cleaning ---
def fetch_stock_data(ticker="TVSMOTOR.NS"):
    """Fetches daily stock data for the last 5 years and cleans it."""
    df = yf.download(tickers=ticker, period="5y", interval="1d")

    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df.reset_index(inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    return df

def fetch_news(symbol="TVS Motor Company"):
    """Fetches news from the last 30 days from NewsAPI."""
    if NEWS_API_KEY == 'YOUR_NEWS_API_KEY':
        print("Warning: NewsAPI key not provided. Skipping news fetching.")
        return pd.DataFrame(columns=["Date", "Headline"])

    to_date = datetime.today().strftime('%Y-%m-%d')
    from_date = (datetime.today() - timedelta(days=30)).strftime('%Y-%m-%d')
    url = f"https://newsapi.org/v2/everything?q=\"{symbol}\"&from={from_date}&to={to_date}&language=en&sortBy=publishedAt&apiKey={NEWS_API_KEY}"
    try:
        res = requests.get(url).json()
        if res.get("status") == "ok":
            articles = res.get("articles", [])
            headlines = [{"Date": a['publishedAt'][:10], "Headline": a['title']} for a in articles]
            return pd.DataFrame(headlines)
        else:
            print(f"Error fetching news: {res.get('message')}")
            return pd.DataFrame(columns=["Date", "Headline"])
    except Exception as e:
        print(f"An exception occurred during news fetching: {e}")
        return pd.DataFrame(columns=["Date", "Headline"])

def fetch_real_events(ticker_symbol, stock_df):
    """
    Fetches real financial and company-specific events relevant to the Indian market.
    """
    all_events = []

    # 1. Fetch Earnings Dates
    try:
        ticker_obj = yf.Ticker(ticker_symbol)
        earnings_dates = ticker_obj.get_earnings_dates(limit=20)
        if earnings_dates is not None and not earnings_dates.empty:
            earnings_df = pd.DataFrame({
                'Date': pd.to_datetime(earnings_dates.index.date),
                'Event_Type': 'Earnings'
            })
            all_events.append(earnings_df)
    except Exception as e:
        print(f"Could not fetch earnings dates: {e}")

    # 2. Add RBI MPC Meeting Dates (Macro-economic factor)
    rbi_mpc_dates = [
        '2020-02-06', '2020-03-27', '2020-05-22', '2020-08-06', '2020-10-09', '2020-12-04',
        '2021-02-05', '2021-04-07', '2021-06-04', '2021-08-06', '2021-10-08', '2021-12-08',
        '2022-02-10', '2022-04-08', '2022-05-04', '2022-06-08', '2022-08-05', '2022-09-30', '2022-12-07',
        '2023-02-08', '2023-04-06', '2023-06-08', '2023-08-10', '2023-10-06', '2023-12-08',
        '2024-02-08', '2024-04-05', '2024-06-07', '2024-08-08', '2024-10-09', '2024-12-06',
        '2025-02-07', '2025-04-04', '2025-06-06', '2025-08-08'
    ]
    rbi_df = pd.DataFrame({ 'Date': pd.to_datetime(rbi_mpc_dates), 'Event_Type': 'RBI MPC Meeting' })
    all_events.append(rbi_df)

    # 3. Add company-specific events for TVS Motor
    if ticker_symbol == 'TVSMOTOR.NS':
        tvs_event_dates = [
            '2020-07-28', # AGM 2020
            '2021-09-16', # Launch of TVS Raider 125
            '2022-05-18', # Launch of updated iQube EV
            '2022-07-06', # Launch of TVS Ronin
            '2023-08-23', # Launch of TVS Apache RTR 310
            '2024-07-25', # Hypothetical future AGM date
            '2025-01-15'  # Hypothetical future EV launch
        ]
        tvs_events_df = pd.DataFrame({
            'Date': pd.to_datetime(tvs_event_dates),
            'Event_Type': 'TVS Product/Corp Event'
        })
        all_events.append(tvs_events_df)

    if not all_events:
        return pd.DataFrame(columns=['Date'])

    events_df = pd.concat(all_events, ignore_index=True)
    min_date, max_date = stock_df['Date'].min(), stock_df['Date'].max()
    events_df = events_df[(events_df['Date'] >= min_date) & (events_df['Date'] <= max_date)]
    return events_df.drop_duplicates(subset=['Date', 'Event_Type']).reset_index(drop=True)


# --- Load Data ---
TICKER = "TVSMOTOR.NS"
COMPANY_NAME = "TVS Motor Company"
stock_df = fetch_stock_data(TICKER)
news_df = fetch_news(COMPANY_NAME)
events_df = fetch_real_events(TICKER, stock_df)


# --- FinBERT Sentiment Scoring ---
if not news_df.empty:
    try:
        tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
        model_fbert = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone").to(DEVICE)

        def get_sentiment(text):
            if not text or not isinstance(text, str): return 0.0
            inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True).to(DEVICE)
            with torch.no_grad():
                outputs = model_fbert(**inputs)
            scores = torch.nn.functional.softmax(outputs.logits, dim=1).cpu().numpy()[0]
            return scores[1] - scores[0] # Positive - Negative

        tqdm.pandas(desc="Calculating Sentiment")
        news_df['Sentiment'] = news_df['Headline'].progress_apply(get_sentiment)
        sentiment_df = news_df.groupby("Date")["Sentiment"].mean().reset_index()
        sentiment_df['Date'] = pd.to_datetime(sentiment_df['Date'])
    except Exception as e:
        print(f"Error loading FinBERT model or processing sentiment: {e}")
        sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])
else:
    sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])

# --- Feature Engineering ---
def add_features(df):
    df['Return'] = df['Close'].pct_change()
    df['MA10'] = df['Close'].rolling(window=10).mean()
    df['MA50'] = df['Close'].rolling(window=50).mean()
    delta = df['Close'].diff()
    up, down = delta.clip(lower=0), -1 * delta.clip(upper=0)
    ema_up, ema_down = up.ewm(com=13, adjust=False).mean(), down.ewm(com=13, adjust=False).mean()
    rs = ema_up / ema_down
    df['RSI'] = 100 - (100 / (1 + rs))
    df['MACD'] = df['Close'].ewm(span=12, adjust=False).mean() - df['Close'].ewm(span=26, adjust=False).mean()
    df['Volatility'] = df['Return'].rolling(window=21).std() * np.sqrt(252)
    return df

stock_df = add_features(stock_df)

# --- Merge Data ---
if not events_df.empty:
    event_dummies = pd.get_dummies(events_df['Event_Type'], prefix='Event', dtype=float)
    events_df = pd.concat([events_df['Date'], event_dummies], axis=1)
    events_df = events_df.groupby('Date').sum().reset_index()
    stock_df = pd.merge(stock_df, events_df, on='Date', how='left')

if not sentiment_df.empty:
    stock_df = pd.merge(stock_df, sentiment_df, on='Date', how='left')
else:
    stock_df['Sentiment'] = 0.0

event_cols = [col for col in stock_df.columns if 'Event_' in col]

# --- Preprocessing ---
stock_df['Target'] = stock_df['Return'].shift(-1)
stock_df['Sentiment'].fillna(0, inplace=True)
for col in event_cols:
    stock_df[col].fillna(0, inplace=True)

tech_cols = ['Return', 'MA10', 'MA50', 'RSI', 'MACD', 'Volatility']
stock_df[tech_cols] = stock_df[tech_cols].fillna(method='ffill').fillna(method='bfill')
stock_df.replace([np.inf, -np.inf], 0, inplace=True)
stock_df.dropna(inplace=True)

# --- Normalize & Split Data ---
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA10', 'MA50', 'RSI', 'MACD', 'Return', 'Volatility', 'Sentiment'] + event_cols

scaler = MinMaxScaler()
stock_df_scaled = stock_df.copy()
stock_df_scaled[features] = scaler.fit_transform(stock_df_scaled[features])

train_size = int(len(stock_df_scaled) * 0.85)
train_df = stock_df_scaled.iloc[:train_size].copy()
test_df = stock_df_scaled.iloc[train_size:].copy()
test_df_unscaled = stock_df.iloc[train_size:].copy()


# --- PyTorch Dataset ---
class StockDataset(Dataset):
    def __init__(self, df, df_unscaled, features, target_col, seq_len):
        self.X, self.y = [], []
        self.close_prices_for_eval = df_unscaled['Close'].values[seq_len-1:-1]

        feature_data = df[features].values
        target_data = df[target_col].values

        for i in range(seq_len, len(df)):
            self.X.append(feature_data[i-seq_len:i])
            self.y.append(target_data[i])

        self.X = np.array(self.X)
        self.y = np.array(self.y)

    def __len__(self): return len(self.X)
    def __getitem__(self, idx):
        return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)

train_dataset = StockDataset(train_df, train_df, features, 'Target', SEQ_LEN)
test_dataset = StockDataset(test_df, test_df_unscaled, features, 'Target', SEQ_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# --- Model Definitions ---
class TLSTM_Attention(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(TLSTM_Attention, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.attention = nn.Linear(hidden_size, 1)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
        context = torch.sum(attn_weights * lstm_out, dim=1)
        return self.fc(context)

class GRU_Model(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(GRU_Model, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.gru(x)
        return self.fc(out[:, -1, :])

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.attn = nn.MultiheadAttention(embed_dim, num_heads, batch_first=True, dropout=rate)
        self.ffn = nn.Sequential(nn.Linear(embed_dim, ff_dim), nn.ReLU(), nn.Linear(ff_dim, embed_dim))
        self.layernorm1, self.layernorm2 = nn.LayerNorm(embed_dim, eps=1e-6), nn.LayerNorm(embed_dim, eps=1e-6)
        self.dropout1, self.dropout2 = nn.Dropout(rate), nn.Dropout(rate)
        self.fc_out = nn.Linear(embed_dim, 1)
    def forward(self, x):
        attn_output, _ = self.attn(x, x, x)
        out1 = self.layernorm1(x + self.dropout1(attn_output))
        ffn_output = self.ffn(out1)
        out2 = self.layernorm2(out1 + self.dropout2(ffn_output))
        return self.fc_out(out2.mean(dim=1))

# --- Training and Evaluation ---
def train_model(model, train_loader, epochs, lr):
    loss_fn = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.train()
    print(f"\n--- Training {model.__class__.__name__} ---")
    for epoch in range(epochs):
        total_loss = 0
        for X_batch, y_batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
            optimizer.zero_grad()
            preds = model(X_batch).squeeze(-1)
            loss = loss_fn(preds, y_batch)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch+1}/{epochs} - Loss: {total_loss / len(train_loader):.8f}")
    return model

def get_predictions(model, loader):
    model.eval()
    preds = []
    with torch.no_grad():
        for X_batch, _ in loader:
            preds.append(model(X_batch.to(DEVICE)).squeeze(-1).cpu().numpy())
    return np.concatenate(preds)

# --- Instantiate and Train All Models ---
input_size = len(features)
hidden_size = 64

# Dynamic head count for Transformer
if input_size % 2 == 0: num_heads = 2
elif input_size % 3 == 0: num_heads = 3
else: num_heads = 1

print(f"Total input features: {input_size}. Using {num_heads} attention heads for Transformer.")

tlstm_model = train_model(TLSTM_Attention(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
gru_model = train_model(GRU_Model(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
trans_model = train_model(TransformerBlock(input_size, num_heads, 64).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)

# --- Ensemble Prediction ---
tlstm_preds = get_predictions(tlstm_model, test_loader)
gru_preds = get_predictions(gru_model, test_loader)
trans_preds = get_predictions(trans_model, test_loader)
final_predicted_returns = (tlstm_preds + gru_preds + trans_preds) / 3.0

# --- Convert Predicted Returns back to Prices for Evaluation ---
last_actual_prices = test_dataset.close_prices_for_eval
predicted_prices = last_actual_prices * (1 + final_predicted_returns)
actual_prices = test_df_unscaled['Close'].values[SEQ_LEN:]

# --- Evaluation ---
print("\n--- Final Ensemble Model Evaluation ---")
mae = mean_absolute_error(actual_prices, predicted_prices)
mse = mean_squared_error(actual_prices, predicted_prices)
r2 = r2_score(actual_prices, predicted_prices)
print(f"Mean Absolute Error (MAE): ₹{mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R²): {r2:.4f}")

# --- Plot Results ---
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(15, 7))
plt.plot(test_df_unscaled['Date'][SEQ_LEN:], actual_prices, label='Actual Price', color='royalblue', linewidth=2)
plt.plot(test_df_unscaled['Date'][SEQ_LEN:], predicted_prices, label='Predicted Price', color='crimson', linestyle='--')
plt.title(f"{TICKER} Stock Price Prediction: Actual vs. Predicted (Ensemble)", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Close Price (INR)", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# NEW: Install necessary libraries in the Colab environment
# !pip install yfinance transformers torch scikit-learn requests tqdm

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from datetime import datetime, timedelta
from tqdm import tqdm
import requests
import yfinance as yf
import warnings

warnings.filterwarnings('ignore')

# --- Constants ---
SEQ_LEN = 30
BATCH_SIZE = 32
EPOCHS = 30
LEARNING_RATE = 0.0001
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
# --- IMPORTANT: Replace with your own NewsAPI key ---
NEWS_API_KEY = '99d1937942dd430cb7e67bbd39c0a713' # <--- REPLACE THIS

print(f"Using device: {DEVICE}")

# --- Data Fetching & Cleaning ---
def fetch_stock_data(ticker="ASHOKLEY.NS"):
    """Fetches daily stock data for the last 5 years and cleans it."""
    df = yf.download(tickers=ticker, period="5y", interval="1d")

    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df.reset_index(inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    return df

def fetch_news(symbol="Ashok Leyland"):
    """Fetches news from the last 30 days from NewsAPI."""
    if NEWS_API_KEY == 'YOUR_NEWS_API_KEY':
        print("Warning: NewsAPI key not provided. Skipping news fetching.")
        return pd.DataFrame(columns=["Date", "Headline"])

    to_date = datetime.today().strftime('%Y-%m-%d')
    from_date = (datetime.today() - timedelta(days=30)).strftime('%Y-%m-%d')
    url = f"https://newsapi.org/v2/everything?q=\"{symbol}\"&from={from_date}&to={to_date}&language=en&sortBy=publishedAt&apiKey={NEWS_API_KEY}"
    try:
        res = requests.get(url).json()
        if res.get("status") == "ok":
            articles = res.get("articles", [])
            headlines = [{"Date": a['publishedAt'][:10], "Headline": a['title']} for a in articles]
            return pd.DataFrame(headlines)
        else:
            print(f"Error fetching news: {res.get('message')}")
            return pd.DataFrame(columns=["Date", "Headline"])
    except Exception as e:
        print(f"An exception occurred during news fetching: {e}")
        return pd.DataFrame(columns=["Date", "Headline"])

def fetch_real_events(ticker_symbol, stock_df):
    """
    Fetches real financial and company-specific events relevant to the Indian market.
    """
    all_events = []

    # 1. Fetch Earnings Dates
    try:
        ticker_obj = yf.Ticker(ticker_symbol)
        earnings_dates = ticker_obj.get_earnings_dates(limit=20)
        if earnings_dates is not None and not earnings_dates.empty:
            earnings_df = pd.DataFrame({
                'Date': pd.to_datetime(earnings_dates.index.date),
                'Event_Type': 'Earnings'
            })
            all_events.append(earnings_df)
    except Exception as e:
        print(f"Could not fetch earnings dates: {e}")

    # 2. Add RBI MPC Meeting Dates (Macro-economic factor)
    rbi_mpc_dates = [
        '2020-02-06', '2020-03-27', '2020-05-22', '2020-08-06', '2020-10-09', '2020-12-04',
        '2021-02-05', '2021-04-07', '2021-06-04', '2021-08-06', '2021-10-08', '2021-12-08',
        '2022-02-10', '2022-04-08', '2022-05-04', '2022-06-08', '2022-08-05', '2022-09-30', '2022-12-07',
        '2023-02-08', '2023-04-06', '2023-06-08', '2023-08-10', '2023-10-06', '2023-12-08',
        '2024-02-08', '2024-04-05', '2024-06-07', '2024-08-08', '2024-10-09', '2024-12-06',
        '2025-02-07', '2025-04-04', '2025-06-06', '2025-08-08'
    ]
    rbi_df = pd.DataFrame({ 'Date': pd.to_datetime(rbi_mpc_dates), 'Event_Type': 'RBI MPC Meeting' })
    all_events.append(rbi_df)

    # 3. Add company-specific events for Ashok Leyland
    if ticker_symbol == 'ASHOKLEY.NS':
        ashokley_event_dates = [
            '2020-09-07', # Launch of AVTR modular truck platform
            '2021-07-28', # AGM 2021
            '2022-01-12', # Launch of new CNG truck range
            '2022-09-05', # Unveiling of new EV buses
            '2023-07-26', # AGM 2023
            '2024-07-30', # Hypothetical future AGM date
            '2025-02-01'  # Hypothetical future product launch
        ]
        ashokley_events_df = pd.DataFrame({
            'Date': pd.to_datetime(ashokley_event_dates),
            'Event_Type': 'AshokLeyland Product/Corp Event'
        })
        all_events.append(ashokley_events_df)

    if not all_events:
        return pd.DataFrame(columns=['Date'])

    events_df = pd.concat(all_events, ignore_index=True)
    min_date, max_date = stock_df['Date'].min(), stock_df['Date'].max()
    events_df = events_df[(events_df['Date'] >= min_date) & (events_df['Date'] <= max_date)]
    return events_df.drop_duplicates(subset=['Date', 'Event_Type']).reset_index(drop=True)


# --- Load Data ---
TICKER = "ASHOKLEY.NS"
COMPANY_NAME = "Ashok Leyland"
stock_df = fetch_stock_data(TICKER)
news_df = fetch_news(COMPANY_NAME)
events_df = fetch_real_events(TICKER, stock_df)


# --- FinBERT Sentiment Scoring ---
if not news_df.empty:
    try:
        tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
        model_fbert = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone").to(DEVICE)

        def get_sentiment(text):
            if not text or not isinstance(text, str): return 0.0
            inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True).to(DEVICE)
            with torch.no_grad():
                outputs = model_fbert(**inputs)
            scores = torch.nn.functional.softmax(outputs.logits, dim=1).cpu().numpy()[0]
            return scores[1] - scores[0]

        tqdm.pandas(desc="Calculating Sentiment")
        news_df['Sentiment'] = news_df['Headline'].progress_apply(get_sentiment)
        sentiment_df = news_df.groupby("Date")["Sentiment"].mean().reset_index()
        sentiment_df['Date'] = pd.to_datetime(sentiment_df['Date'])
    except Exception as e:
        print(f"Error loading FinBERT model or processing sentiment: {e}")
        sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])
else:
    sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])

# --- Feature Engineering ---
def add_features(df):
    df['Return'] = df['Close'].pct_change()
    df['MA10'] = df['Close'].rolling(window=10).mean()
    df['MA50'] = df['Close'].rolling(window=50).mean()
    delta = df['Close'].diff()
    up, down = delta.clip(lower=0), -1 * delta.clip(upper=0)
    ema_up, ema_down = up.ewm(com=13, adjust=False).mean(), down.ewm(com=13, adjust=False).mean()
    rs = ema_up / ema_down
    df['RSI'] = 100 - (100 / (1 + rs))
    df['MACD'] = df['Close'].ewm(span=12, adjust=False).mean() - df['Close'].ewm(span=26, adjust=False).mean()
    df['Volatility'] = df['Return'].rolling(window=21).std() * np.sqrt(252)
    return df

stock_df = add_features(stock_df)

# --- Merge Data ---
if not events_df.empty:
    event_dummies = pd.get_dummies(events_df['Event_Type'], prefix='Event', dtype=float)
    events_df = pd.concat([events_df['Date'], event_dummies], axis=1)
    events_df = events_df.groupby('Date').sum().reset_index()
    stock_df = pd.merge(stock_df, events_df, on='Date', how='left')

if not sentiment_df.empty:
    stock_df = pd.merge(stock_df, sentiment_df, on='Date', how='left')
else:
    stock_df['Sentiment'] = 0.0

event_cols = [col for col in stock_df.columns if 'Event_' in col]

# --- Preprocessing ---
stock_df['Target'] = stock_df['Return'].shift(-1)
stock_df['Sentiment'].fillna(0, inplace=True)
for col in event_cols:
    stock_df[col].fillna(0, inplace=True)

tech_cols = ['Return', 'MA10', 'MA50', 'RSI', 'MACD', 'Volatility']
stock_df[tech_cols] = stock_df[tech_cols].fillna(method='ffill').fillna(method='bfill')
stock_df.replace([np.inf, -np.inf], 0, inplace=True)
stock_df.dropna(inplace=True)

# --- Normalize & Split Data ---
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA10', 'MA50', 'RSI', 'MACD', 'Return', 'Volatility', 'Sentiment'] + event_cols

scaler = MinMaxScaler()
stock_df_scaled = stock_df.copy()
stock_df_scaled[features] = scaler.fit_transform(stock_df_scaled[features])

train_size = int(len(stock_df_scaled) * 0.85)
train_df = stock_df_scaled.iloc[:train_size].copy()
test_df = stock_df_scaled.iloc[train_size:].copy()
test_df_unscaled = stock_df.iloc[train_size:].copy()


# --- PyTorch Dataset ---
class StockDataset(Dataset):
    def __init__(self, df, df_unscaled, features, target_col, seq_len):
        self.X, self.y = [], []
        self.close_prices_for_eval = df_unscaled['Close'].values[seq_len-1:-1]

        feature_data = df[features].values
        target_data = df[target_col].values

        for i in range(seq_len, len(df)):
            self.X.append(feature_data[i-seq_len:i])
            self.y.append(target_data[i])

        self.X = np.array(self.X)
        self.y = np.array(self.y)

    def __len__(self): return len(self.X)
    def __getitem__(self, idx):
        return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)

train_dataset = StockDataset(train_df, train_df, features, 'Target', SEQ_LEN)
test_dataset = StockDataset(test_df, test_df_unscaled, features, 'Target', SEQ_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# --- Model Definitions ---
class TLSTM_Attention(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(TLSTM_Attention, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.attention = nn.Linear(hidden_size, 1)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
        context = torch.sum(attn_weights * lstm_out, dim=1)
        return self.fc(context)

class GRU_Model(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(GRU_Model, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.gru(x)
        return self.fc(out[:, -1, :])

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.attn = nn.MultiheadAttention(embed_dim, num_heads, batch_first=True, dropout=rate)
        self.ffn = nn.Sequential(nn.Linear(embed_dim, ff_dim), nn.ReLU(), nn.Linear(ff_dim, embed_dim))
        self.layernorm1, self.layernorm2 = nn.LayerNorm(embed_dim, eps=1e-6), nn.LayerNorm(embed_dim, eps=1e-6)
        self.dropout1, self.dropout2 = nn.Dropout(rate), nn.Dropout(rate)
        self.fc_out = nn.Linear(embed_dim, 1)
    def forward(self, x):
        attn_output, _ = self.attn(x, x, x)
        out1 = self.layernorm1(x + self.dropout1(attn_output))
        ffn_output = self.ffn(out1)
        out2 = self.layernorm2(out1 + self.dropout2(ffn_output))
        return self.fc_out(out2.mean(dim=1))

# --- Training and Evaluation ---
def train_model(model, train_loader, epochs, lr):
    loss_fn = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.train()
    print(f"\n--- Training {model.__class__.__name__} ---")
    for epoch in range(epochs):
        total_loss = 0
        for X_batch, y_batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
            optimizer.zero_grad()
            preds = model(X_batch).squeeze(-1)
            loss = loss_fn(preds, y_batch)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch+1}/{epochs} - Loss: {total_loss / len(train_loader):.8f}")
    return model

def get_predictions(model, loader):
    model.eval()
    preds = []
    with torch.no_grad():
        for X_batch, _ in loader:
            preds.append(model(X_batch.to(DEVICE)).squeeze(-1).cpu().numpy())
    return np.concatenate(preds)

# --- Instantiate and Train All Models ---
input_size = len(features)
hidden_size = 64

# Dynamic head count for Transformer
if input_size % 2 == 0: num_heads = 2
elif input_size % 3 == 0: num_heads = 3
else: num_heads = 1

print(f"Total input features: {input_size}. Using {num_heads} attention heads for Transformer.")

tlstm_model = train_model(TLSTM_Attention(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
gru_model = train_model(GRU_Model(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
trans_model = train_model(TransformerBlock(input_size, num_heads, 64).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)

# --- Ensemble Prediction ---
tlstm_preds = get_predictions(tlstm_model, test_loader)
gru_preds = get_predictions(gru_model, test_loader)
trans_preds = get_predictions(trans_model, test_loader)
final_predicted_returns = (tlstm_preds + gru_preds + trans_preds) / 3.0

# --- Convert Predicted Returns back to Prices for Evaluation ---
last_actual_prices = test_dataset.close_prices_for_eval
predicted_prices = last_actual_prices * (1 + final_predicted_returns)
actual_prices = test_df_unscaled['Close'].values[SEQ_LEN:]

# --- Evaluation ---
print("\n--- Final Ensemble Model Evaluation ---")
mae = mean_absolute_error(actual_prices, predicted_prices)
mse = mean_squared_error(actual_prices, predicted_prices)
r2 = r2_score(actual_prices, predicted_prices)
print(f"Mean Absolute Error (MAE): ₹{mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R²): {r2:.4f}")

# --- Plot Results ---
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(15, 7))
plt.plot(test_df_unscaled['Date'][SEQ_LEN:], actual_prices, label='Actual Price', color='royalblue', linewidth=2)
plt.plot(test_df_unscaled['Date'][SEQ_LEN:], predicted_prices, label='Predicted Price', color='crimson', linestyle='--')
plt.title(f"{TICKER} Stock Price Prediction: Actual vs. Predicted (Ensemble)", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Close Price (INR)", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# NEW: Install necessary libraries in the Colab environment
# !pip install yfinance transformers torch scikit-learn requests tqdm

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from datetime import datetime, timedelta
from tqdm import tqdm
import requests
import yfinance as yf
import warnings

warnings.filterwarnings('ignore')

# --- Constants ---
SEQ_LEN = 30
BATCH_SIZE = 32
EPOCHS = 30
LEARNING_RATE = 0.0001
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
# --- IMPORTANT: Replace with your own NewsAPI key ---
NEWS_API_KEY = '99d1937942dd430cb7e67bbd39c0a713' # <--- REPLACE THIS

print(f"Using device: {DEVICE}")

# --- Data Fetching & Cleaning ---
def fetch_stock_data(ticker="MRF.NS"):
    """Fetches daily stock data for the last 5 years and cleans it."""
    df = yf.download(tickers=ticker, period="5y", interval="1d")

    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df.reset_index(inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    return df

def fetch_news(symbol="MRF Tyres"):
    """Fetches news from the last 30 days from NewsAPI."""
    if NEWS_API_KEY == 'YOUR_NEWS_API_KEY':
        print("Warning: NewsAPI key not provided. Skipping news fetching.")
        return pd.DataFrame(columns=["Date", "Headline"])

    to_date = datetime.today().strftime('%Y-%m-%d')
    from_date = (datetime.today() - timedelta(days=30)).strftime('%Y-%m-%d')
    url = f"https://newsapi.org/v2/everything?q=\"{symbol}\"&from={from_date}&to={to_date}&language=en&sortBy=publishedAt&apiKey={NEWS_API_KEY}"
    try:
        res = requests.get(url).json()
        if res.get("status") == "ok":
            articles = res.get("articles", [])
            headlines = [{"Date": a['publishedAt'][:10], "Headline": a['title']} for a in articles]
            return pd.DataFrame(headlines)
        else:
            print(f"Error fetching news: {res.get('message')}")
            return pd.DataFrame(columns=["Date", "Headline"])
    except Exception as e:
        print(f"An exception occurred during news fetching: {e}")
        return pd.DataFrame(columns=["Date", "Headline"])

def fetch_real_events(ticker_symbol, stock_df):
    """
    Fetches real financial and company-specific events relevant to the Indian market.
    """
    all_events = []

    # 1. Fetch Earnings Dates
    try:
        ticker_obj = yf.Ticker(ticker_symbol)
        earnings_dates = ticker_obj.get_earnings_dates(limit=20)
        if earnings_dates is not None and not earnings_dates.empty:
            earnings_df = pd.DataFrame({
                'Date': pd.to_datetime(earnings_dates.index.date),
                'Event_Type': 'Earnings'
            })
            all_events.append(earnings_df)
    except Exception as e:
        print(f"Could not fetch earnings dates: {e}")

    # 2. Add RBI MPC Meeting Dates (Macro-economic factor)
    rbi_mpc_dates = [
        '2020-02-06', '2020-03-27', '2020-05-22', '2020-08-06', '2020-10-09', '2020-12-04',
        '2021-02-05', '2021-04-07', '2021-06-04', '2021-08-06', '2021-10-08', '2021-12-08',
        '2022-02-10', '2022-04-08', '2022-05-04', '2022-06-08', '2022-08-05', '2022-09-30', '2022-12-07',
        '2023-02-08', '2023-04-06', '2023-06-08', '2023-08-10', '2023-10-06', '2023-12-08',
        '2024-02-08', '2024-04-05', '2024-06-07', '2024-08-08', '2024-10-09', '2024-12-06',
        '2025-02-07', '2025-04-04', '2025-06-06', '2025-08-08'
    ]
    rbi_df = pd.DataFrame({ 'Date': pd.to_datetime(rbi_mpc_dates), 'Event_Type': 'RBI MPC Meeting' })
    all_events.append(rbi_df)

    # 3. Add company-specific events for MRF
    if ticker_symbol == 'MRF.NS':
        mrf_event_dates = [
            '2020-08-11', # AGM 2020
            '2021-08-12', # AGM 2021
            '2022-08-10', # AGM 2022
            '2023-06-13', # Stock hits Rs 1 lakh mark
            '2023-08-08', # AGM 2023
            '2024-08-13', # Hypothetical future AGM date
            '2025-04-10'  # Hypothetical future product/sponsorship announcement
        ]
        mrf_events_df = pd.DataFrame({
            'Date': pd.to_datetime(mrf_event_dates),
            'Event_Type': 'MRF Corp/Milestone Event'
        })
        all_events.append(mrf_events_df)

    if not all_events:
        return pd.DataFrame(columns=['Date'])

    events_df = pd.concat(all_events, ignore_index=True)
    min_date, max_date = stock_df['Date'].min(), stock_df['Date'].max()
    events_df = events_df[(events_df['Date'] >= min_date) & (events_df['Date'] <= max_date)]
    return events_df.drop_duplicates(subset=['Date', 'Event_Type']).reset_index(drop=True)


# --- Load Data ---
TICKER = "MRF.NS"
COMPANY_NAME = "MRF Tyres"
stock_df = fetch_stock_data(TICKER)
news_df = fetch_news(COMPANY_NAME)
events_df = fetch_real_events(TICKER, stock_df)


# --- FinBERT Sentiment Scoring ---
if not news_df.empty:
    try:
        tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
        model_fbert = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone").to(DEVICE)

        def get_sentiment(text):
            if not text or not isinstance(text, str): return 0.0
            inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True).to(DEVICE)
            with torch.no_grad():
                outputs = model_fbert(**inputs)
            scores = torch.nn.functional.softmax(outputs.logits, dim=1).cpu().numpy()[0]
            return scores[1] - scores[0]

        tqdm.pandas(desc="Calculating Sentiment")
        news_df['Sentiment'] = news_df['Headline'].progress_apply(get_sentiment)
        sentiment_df = news_df.groupby("Date")["Sentiment"].mean().reset_index()
        sentiment_df['Date'] = pd.to_datetime(sentiment_df['Date'])
    except Exception as e:
        print(f"Error loading FinBERT model or processing sentiment: {e}")
        sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])
else:
    sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])

# --- Feature Engineering ---
def add_features(df):
    df['Return'] = df['Close'].pct_change()
    df['MA10'] = df['Close'].rolling(window=10).mean()
    df['MA50'] = df['Close'].rolling(window=50).mean()
    delta = df['Close'].diff()
    up, down = delta.clip(lower=0), -1 * delta.clip(upper=0)
    ema_up, ema_down = up.ewm(com=13, adjust=False).mean(), down.ewm(com=13, adjust=False).mean()
    rs = ema_up / ema_down
    df['RSI'] = 100 - (100 / (1 + rs))
    df['MACD'] = df['Close'].ewm(span=12, adjust=False).mean() - df['Close'].ewm(span=26, adjust=False).mean()
    df['Volatility'] = df['Return'].rolling(window=21).std() * np.sqrt(252)
    return df

stock_df = add_features(stock_df)

# --- Merge Data ---
if not events_df.empty:
    event_dummies = pd.get_dummies(events_df['Event_Type'], prefix='Event', dtype=float)
    events_df = pd.concat([events_df['Date'], event_dummies], axis=1)
    events_df = events_df.groupby('Date').sum().reset_index()
    stock_df = pd.merge(stock_df, events_df, on='Date', how='left')

if not sentiment_df.empty:
    stock_df = pd.merge(stock_df, sentiment_df, on='Date', how='left')
else:
    stock_df['Sentiment'] = 0.0

event_cols = [col for col in stock_df.columns if 'Event_' in col]

# --- Preprocessing ---
stock_df['Target'] = stock_df['Return'].shift(-1)
stock_df['Sentiment'].fillna(0, inplace=True)
for col in event_cols:
    stock_df[col].fillna(0, inplace=True)

tech_cols = ['Return', 'MA10', 'MA50', 'RSI', 'MACD', 'Volatility']
stock_df[tech_cols] = stock_df[tech_cols].fillna(method='ffill').fillna(method='bfill')
stock_df.replace([np.inf, -np.inf], 0, inplace=True)
stock_df.dropna(inplace=True)

# --- Normalize & Split Data ---
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA10', 'MA50', 'RSI', 'MACD', 'Return', 'Volatility', 'Sentiment'] + event_cols

scaler = MinMaxScaler()
stock_df_scaled = stock_df.copy()
stock_df_scaled[features] = scaler.fit_transform(stock_df_scaled[features])

train_size = int(len(stock_df_scaled) * 0.85)
train_df = stock_df_scaled.iloc[:train_size].copy()
test_df = stock_df_scaled.iloc[train_size:].copy()
test_df_unscaled = stock_df.iloc[train_size:].copy()


# --- PyTorch Dataset ---
class StockDataset(Dataset):
    def __init__(self, df, df_unscaled, features, target_col, seq_len):
        self.X, self.y = [], []
        self.close_prices_for_eval = df_unscaled['Close'].values[seq_len-1:-1]

        feature_data = df[features].values
        target_data = df[target_col].values

        for i in range(seq_len, len(df)):
            self.X.append(feature_data[i-seq_len:i])
            self.y.append(target_data[i])

        self.X = np.array(self.X)
        self.y = np.array(self.y)

    def __len__(self): return len(self.X)
    def __getitem__(self, idx):
        return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)

train_dataset = StockDataset(train_df, train_df, features, 'Target', SEQ_LEN)
test_dataset = StockDataset(test_df, test_df_unscaled, features, 'Target', SEQ_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# --- Model Definitions ---
class TLSTM_Attention(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(TLSTM_Attention, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.attention = nn.Linear(hidden_size, 1)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
        context = torch.sum(attn_weights * lstm_out, dim=1)
        return self.fc(context)

class GRU_Model(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(GRU_Model, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.gru(x)
        return self.fc(out[:, -1, :])

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.attn = nn.MultiheadAttention(embed_dim, num_heads, batch_first=True, dropout=rate)
        self.ffn = nn.Sequential(nn.Linear(embed_dim, ff_dim), nn.ReLU(), nn.Linear(ff_dim, embed_dim))
        self.layernorm1, self.layernorm2 = nn.LayerNorm(embed_dim, eps=1e-6), nn.LayerNorm(embed_dim, eps=1e-6)
        self.dropout1, self.dropout2 = nn.Dropout(rate), nn.Dropout(rate)
        self.fc_out = nn.Linear(embed_dim, 1)
    def forward(self, x):
        attn_output, _ = self.attn(x, x, x)
        out1 = self.layernorm1(x + self.dropout1(attn_output))
        ffn_output = self.ffn(out1)
        out2 = self.layernorm2(out1 + self.dropout2(ffn_output))
        return self.fc_out(out2.mean(dim=1))

# --- Training and Evaluation ---
def train_model(model, train_loader, epochs, lr):
    loss_fn = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.train()
    print(f"\n--- Training {model.__class__.__name__} ---")
    for epoch in range(epochs):
        total_loss = 0
        for X_batch, y_batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
            optimizer.zero_grad()
            preds = model(X_batch).squeeze(-1)
            loss = loss_fn(preds, y_batch)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch+1}/{epochs} - Loss: {total_loss / len(train_loader):.8f}")
    return model

def get_predictions(model, loader):
    model.eval()
    preds = []
    with torch.no_grad():
        for X_batch, _ in loader:
            preds.append(model(X_batch.to(DEVICE)).squeeze(-1).cpu().numpy())
    return np.concatenate(preds)

# --- Instantiate and Train All Models ---
input_size = len(features)
hidden_size = 64

# Dynamic head count for Transformer
if input_size % 2 == 0: num_heads = 2
elif input_size % 3 == 0: num_heads = 3
else: num_heads = 1

print(f"Total input features: {input_size}. Using {num_heads} attention heads for Transformer.")

tlstm_model = train_model(TLSTM_Attention(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
gru_model = train_model(GRU_Model(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
trans_model = train_model(TransformerBlock(input_size, num_heads, 64).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)

# --- Ensemble Prediction ---
tlstm_preds = get_predictions(tlstm_model, test_loader)
gru_preds = get_predictions(gru_model, test_loader)
trans_preds = get_predictions(trans_model, test_loader)
final_predicted_returns = (tlstm_preds + gru_preds + trans_preds) / 3.0

# --- Convert Predicted Returns back to Prices for Evaluation ---
last_actual_prices = test_dataset.close_prices_for_eval
predicted_prices = last_actual_prices * (1 + final_predicted_returns)
actual_prices = test_df_unscaled['Close'].values[SEQ_LEN:]

# --- Evaluation ---
print("\n--- Final Ensemble Model Evaluation ---")
mae = mean_absolute_error(actual_prices, predicted_prices)
mse = mean_squared_error(actual_prices, predicted_prices)
r2 = r2_score(actual_prices, predicted_prices)
print(f"Mean Absolute Error (MAE): ₹{mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R²): {r2:.4f}")

# --- Plot Results ---
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(15, 7))
plt.plot(test_df_unscaled['Date'][SEQ_LEN:], actual_prices, label='Actual Price', color='royalblue', linewidth=2)
plt.plot(test_df_unscaled['Date'][SEQ_LEN:], predicted_prices, label='Predicted Price', color='crimson', linestyle='--')
plt.title(f"{TICKER} Stock Price Prediction: Actual vs. Predicted (Ensemble)", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Close Price (INR)", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Real-time next-day prediction refactor of your notebook
# Run in Colab (GPU recommended). Replace NEWS_API_KEY if you want full NewsAPI access.

!pip install -q yfinance transformers torch scikit-learn requests tqdm

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
from torch.utils.data import Dataset, DataLoader
from datetime import datetime, timedelta
from tqdm import tqdm
import requests
import yfinance as yf
import warnings
from pandas.tseries.offsets import BDay

warnings.filterwarnings('ignore')

# --- Constants ---
SEQ_LEN = 30
BATCH_SIZE = 32
EPOCHS = 15          # Lowered for faster runs; increase as needed
LEARNING_RATE = 1e-4
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
NEWS_API_KEY = '99d1937942dd430cb7e67bbd39c0a713'  # replace if you have another key
USE_FINBERT = False   # set True if you want to use FinBERT (heavier)
NEWS_DAYS = 7         # fetch recent N days of news for realtime sentiment
TICKER = "RELIANCE.NS"
COMPANY_NAME = "Reliance Industries"

print(f"Using device: {DEVICE}")

# ---------------- Data fetching ----------------
def fetch_stock_data(ticker="RELIANCE.NS", period="5y"):
    df = yf.download(tickers=ticker, period=period, interval="1d", progress=False)
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)
    df.reset_index(inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    return df

def fetch_news_recent(query=COMPANY_NAME, days=NEWS_DAYS, api_key=NEWS_API_KEY):
    # returns DataFrame with Date (YYYY-MM-DD) and Headline
    if not api_key or api_key == 'YOUR_NEWS_API_KEY':
        print("No NewsAPI key provided or invalid — skipping NewsAPI. Empty news returned.")
        return pd.DataFrame(columns=["Date","Headline"])
    to_date = datetime.utcnow().date()
    from_date = to_date - timedelta(days=days)
    url = ("https://newsapi.org/v2/everything?"
           f"q={requests.utils.quote(query)}&from={from_date}&to={to_date}"
           "&language=en&sortBy=publishedAt&pageSize=100&apiKey=" + api_key)
    try:
        res = requests.get(url, timeout=15).json()
        if res.get("status") == "ok":
            articles = res.get("articles", [])
            rows = [{"Date": a['publishedAt'][:10], "Headline": a['title']} for a in articles if a.get('title')]
            return pd.DataFrame(rows)
        else:
            print("NewsAPI error:", res.get("message"))
            return pd.DataFrame(columns=["Date","Headline"])
    except Exception as e:
        print("News fetch exception:", e)
        return pd.DataFrame(columns=["Date","Headline"])

# ---------------- Events fetching ----------------
def fetch_events_for_range(ticker_symbol, stock_df):
    all_events = []
    # 1) Earnings (yfinance)
    try:
        ticker_obj = yf.Ticker(ticker_symbol)
        # get_earnings_dates returns DataFrame with index of datetimes for some versions
        ed = ticker_obj.get_earnings_dates(limit=20)
        if ed is not None and not ed.empty:
            # normalize depending on object structure
            if isinstance(ed, pd.DataFrame):
                try:
                    ed_index = pd.to_datetime(ed.index.date)
                except Exception:
                    ed_index = pd.to_datetime(ed['Earnings Date']) if 'Earnings Date' in ed.columns else pd.to_datetime(ed.index)
                earnings_df = pd.DataFrame({'Date': ed_index, 'Event_Type': 'Earnings'})
                all_events.append(earnings_df)
    except Exception as e:
        print("Earnings fetch failed:", e)

    # 2) RBI MPC Dates (static list; keep updated manually)
    rbi_mpc_dates = [
        '2020-02-06','2020-03-27','2020-05-22','2020-08-06','2020-10-09','2020-12-04',
        '2021-02-05','2021-04-07','2021-06-04','2021-08-06','2021-10-08','2021-12-08',
        '2022-02-10','2022-04-08','2022-05-04','2022-06-08','2022-08-05','2022-09-30','2022-12-07',
        '2023-02-08','2023-04-06','2023-06-08','2023-08-10','2023-10-06','2023-12-08',
        '2024-02-08','2024-04-05','2024-06-07','2024-08-08','2024-10-09','2024-12-06',
        '2025-02-07','2025-04-04','2025-06-06','2025-08-08','2025-10-09','2025-12-05'
    ]
    rbi_df = pd.DataFrame({'Date': pd.to_datetime(rbi_mpc_dates), 'Event_Type': 'RBI MPC Meeting'})
    all_events.append(rbi_df)

    # 3) Company AGMs / other known events (extendable)
    if ticker_symbol.upper() == 'RELIANCE.NS':
        reliance_event_dates = [
            '2020-07-15','2021-06-24','2022-08-29','2023-08-28','2024-08-29'
            # don't include hypothetical future unless confirmed
        ]
        rem = pd.DataFrame({'Date': pd.to_datetime(reliance_event_dates), 'Event_Type': 'Reliance AGM'})
        all_events.append(rem)

    if not all_events:
        return pd.DataFrame(columns=['Date'])
    events_df = pd.concat(all_events, ignore_index=True)
    min_date, max_date = stock_df['Date'].min(), stock_df['Date'].max()
    events_df = events_df[(events_df['Date'] >= min_date) & (events_df['Date'] <= max_date)]
    events_df = events_df.drop_duplicates(subset=['Date','Event_Type']).reset_index(drop=True)
    return events_df

# ---------------- Sentiment ----------------
def compute_sentiment(news_df):
    # Returns aggregated daily sentiment (Date as datetime, Sentiment numeric)
    if news_df.empty:
        return pd.DataFrame(columns=['Date','Sentiment'])

    # Choose FinBERT if available and desired
    if USE_FINBERT:
        try:
            tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
            model_fbert = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone").to(DEVICE)
            def f_sent(text):
                if not isinstance(text,str) or not text.strip(): return 0.0
                inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True).to(DEVICE)
                with torch.no_grad():
                    outputs = model_fbert(**inputs)
                scores = torch.nn.functional.softmax(outputs.logits, dim=1).cpu().numpy()[0]
                # The FinBERT variant may have ordering; we use positive - negative as proxy
                # (adjust mapping if model label meanings differ)
                return float(scores[1] - scores[0])
            news_df['Sentiment'] = news_df['Headline'].apply(f_sent)
            out = news_df.groupby('Date', as_index=False)['Sentiment'].mean()
            out['Date'] = pd.to_datetime(out['Date'])
            return out
        except Exception as e:
            print("FinBERT failed, falling back to HF sentiment pipeline:", e)

    # Fallback: HF sentiment pipeline (fast)
    try:
        nlp = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english", device=0 if DEVICE=='cuda' else -1)
        def simple_sent(text):
            if not isinstance(text,str) or not text.strip(): return 0.0
            out = nlp(text[:512])[0]
            # label 'POSITIVE' -> +1, 'NEGATIVE' -> -1 ; if score used
            return (1.0 * out['score']) if out['label']=='POSITIVE' else (-1.0 * out['score'])
        news_df['Sentiment'] = news_df['Headline'].apply(simple_sent)
        out = news_df.groupby('Date', as_index=False)['Sentiment'].mean()
        out['Date'] = pd.to_datetime(out['Date'])
        return out
    except Exception as e:
        print("Fallback sentiment failed:", e)
        return pd.DataFrame(columns=['Date','Sentiment'])

# ---------------- Feature engineering ----------------
def add_features(df):
    df = df.copy()
    df['Return'] = df['Close'].pct_change()
    df['MA10'] = df['Close'].rolling(window=10).mean()
    df['MA50'] = df['Close'].rolling(window=50).mean()
    delta = df['Close'].diff()
    up, down = delta.clip(lower=0), -1 * delta.clip(upper=0)
    ema_up = up.ewm(com=13, adjust=False).mean()
    ema_down = down.ewm(com=13, adjust=False).mean()
    rs = ema_up / (ema_down.replace(0, np.nan))
    df['RSI'] = 100 - (100 / (1 + rs))
    df['MACD'] = df['Close'].ewm(span=12, adjust=False).mean() - df['Close'].ewm(span=26, adjust=False).mean()
    df['Volatility'] = df['Return'].rolling(window=21).std() * np.sqrt(252)
    return df

# ---------------- Load and prepare data ----------------
stock_df = fetch_stock_data(TICKER, period="5y")
news_df = fetch_news_recent(COMPANY_NAME, days=NEWS_DAYS)
events_df = fetch_events_for_range(TICKER, stock_df)

# Sentiment aggregation
sentiment_df = compute_sentiment(news_df)

stock_df = add_features(stock_df)

# Merge events into stock_df as dummies (for dates present in stock_df range)
event_cols = []
if not events_df.empty:
    event_dummies = pd.get_dummies(events_df['Event_Type'], prefix='Event', dtype=float)
    events_df2 = pd.concat([events_df['Date'], event_dummies], axis=1)
    events_df2 = events_df2.groupby('Date').sum().reset_index()
    stock_df = pd.merge(stock_df, events_df2, on='Date', how='left')
    event_cols = [c for c in stock_df.columns if c.startswith('Event_')]
    for c in event_cols:
        stock_df[c].fillna(0.0, inplace=True)

# Merge sentiment
if not sentiment_df.empty:
    stock_df = pd.merge(stock_df, sentiment_df, on='Date', how='left')
    stock_df['Sentiment'].fillna(0.0, inplace=True)
else:
    stock_df['Sentiment'] = 0.0

# Target (next-day return)
stock_df['Target'] = stock_df['Return'].shift(-1)

# clean & fill
tech_cols = ['Return','MA10','MA50','RSI','MACD','Volatility']
for col in tech_cols:
    if col in stock_df.columns:
        stock_df[col] = stock_df[col].fillna(method='ffill').fillna(method='bfill')
stock_df.replace([np.inf, -np.inf], 0, inplace=True)
stock_df.dropna(inplace=True)   # ensure sequences valid

# Features list
features = ['Open','High','Low','Close','Volume','MA10','MA50','RSI','MACD','Return','Volatility','Sentiment'] + event_cols
# drop any feature not present (safety)
features = [f for f in features if f in stock_df.columns]

# ---------------- Train/Test split and scaler (no leakage) ----------------
train_size = int(len(stock_df) * 0.85)
train_df = stock_df.iloc[:train_size].copy()
test_df = stock_df.iloc[train_size:].copy()
test_df_unscaled = stock_df.iloc[train_size:].copy()

scaler = MinMaxScaler()
# Fit scaler on train only
train_features = scaler.fit_transform(train_df[features])
train_df_scaled = train_df.copy()
train_df_scaled[features] = train_features

# transform test using same scaler
test_df_scaled = test_df.copy()
test_df_scaled[features] = scaler.transform(test_df[features])

# ---------------- Dataset class ----------------
class StockDataset(Dataset):
    def __init__(self, df_scaled, df_unscaled, features, target_col, seq_len):
        self.X = []
        self.y = []
        self.close_prices_for_eval = df_unscaled['Close'].values[seq_len-1:-1]  # used later
        feat = df_scaled[features].values
        targ = df_scaled[target_col].values if target_col in df_scaled.columns else df_unscaled[target_col].values
        for i in range(seq_len, len(df_scaled)):
            self.X.append(feat[i-seq_len:i])
            self.y.append(targ[i])
        self.X = np.array(self.X)
        self.y = np.array(self.y)
    def __len__(self): return len(self.X)
    def __getitem__(self, idx):
        return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)

# build datasets
train_dataset = StockDataset(train_df_scaled, train_df, features, 'Target', SEQ_LEN)
test_dataset = StockDataset(test_df_scaled, test_df_unscaled, features, 'Target', SEQ_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# ---------------- Models ----------------
class TLSTM_Attention(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(TLSTM_Attention, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.attention = nn.Linear(hidden_size, 1)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
        context = torch.sum(attn_weights * lstm_out, dim=1)
        return self.fc(context)

class GRU_Model(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(GRU_Model, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.gru(x)
        return self.fc(out[:, -1, :])

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.attn = nn.MultiheadAttention(embed_dim, num_heads, batch_first=True, dropout=rate)
        self.ffn = nn.Sequential(nn.Linear(embed_dim, ff_dim), nn.ReLU(), nn.Linear(ff_dim, embed_dim))
        self.layernorm1, self.layernorm2 = nn.LayerNorm(embed_dim, eps=1e-6), nn.LayerNorm(embed_dim, eps=1e-6)
        self.dropout1, self.dropout2 = nn.Dropout(rate), nn.Dropout(rate)
        self.fc_out = nn.Linear(embed_dim, 1)
    def forward(self, x):
        attn_output, _ = self.attn(x, x, x)
        out1 = self.layernorm1(x + self.dropout1(attn_output))
        ffn_output = self.ffn(out1)
        out2 = self.layernorm2(out1 + self.dropout2(ffn_output))
        return self.fc_out(out2.mean(dim=1))

# ---------------- Training/evaluation helpers ----------------
def train_model(model, train_loader, epochs, lr):
    loss_fn = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.to(DEVICE)
    model.train()
    print(f"\n--- Training {model.__class__.__name__} ---")
    for epoch in range(epochs):
        total_loss = 0.0
        for X_batch, y_batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}", leave=False):
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
            optimizer.zero_grad()
            preds = model(X_batch).squeeze(-1)
            loss = loss_fn(preds, y_batch)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        avg = total_loss / max(len(train_loader),1)
        print(f"Epoch {epoch+1}/{epochs} - Loss: {avg:.8f}")
    return model

def get_predictions(model, loader):
    model.eval()
    preds = []
    with torch.no_grad():
        for X_batch, _ in loader:
            X_batch = X_batch.to(DEVICE)
            out = model(X_batch).squeeze(-1).cpu().numpy()
            preds.append(out)
    if len(preds)==0:
        return np.array([])
    return np.concatenate(preds)

# ---------------- Train ensemble ----------------
input_size = len(features)
hidden_size = 64
if input_size % 2 == 0:
    num_heads = 2
elif input_size % 3 == 0:
    num_heads = 3
else:
    num_heads = 1

print(f"Input features: {input_size}. Using {num_heads} heads for Transformer.")

tlstm_model = TLSTM_Attention(input_size, hidden_size)
gru_model = GRU_Model(input_size, hidden_size)
trans_model = TransformerBlock(input_size, num_heads, ff_dim=64)

tlstm_model = train_model(tlstm_model, train_loader, EPOCHS, LEARNING_RATE)
gru_model = train_model(gru_model, train_loader, EPOCHS, LEARNING_RATE)
trans_model = train_model(trans_model, train_loader, EPOCHS, LEARNING_RATE)

# Ensemble predictions on test (for reporting)
tlstm_preds = get_predictions(tlstm_model, test_loader)
gru_preds = get_predictions(gru_model, test_loader)
trans_preds = get_predictions(trans_model, test_loader)
if len(tlstm_preds) and len(gru_preds) and len(trans_preds):
    final_predicted_returns = (tlstm_preds + gru_preds + trans_preds) / 3.0
    last_actual_prices = test_dataset.close_prices_for_eval
    predicted_prices = last_actual_prices * (1 + final_predicted_returns)
    actual_prices = test_df_unscaled['Close'].values[SEQ_LEN:]
    print("\n--- Ensemble Test Evaluation ---")
    print(f"MAE: ₹{mean_absolute_error(actual_prices, predicted_prices):.2f}")
    print(f"MSE: {mean_squared_error(actual_prices, predicted_prices):.2f}")
    print(f"R2: {r2_score(actual_prices, predicted_prices):.4f}")

# ---------------- Predict next trading day (realtime) ----------------
def predict_next_day(models, scaler, stock_df_full, features, seq_len=SEQ_LEN):
    """
    models: list or tuple of trained models [tlstm, gru, trans]
    scaler: MinMaxScaler fitted on TRAIN data
    stock_df_full: full stock_df used earlier (unscaled) - must include latest date rows
    features: list of feature col names (must exist in stock_df_full)
    returns: dict with predicted_return, predicted_price, next_trade_date
    """
    # ensure latest available date in stock_df is a trading day; next trading day:
    last_date = stock_df_full['Date'].max()
    next_trade = (last_date + BDay(1)).date()

    # prepare the most recent seq_len rows (unscaled), compute features on latest slice
    recent = stock_df_full.copy().sort_values('Date').reset_index(drop=True)
    # verify we have at least seq_len rows
    if len(recent) < seq_len:
        raise ValueError("Not enough rows to form sequence for prediction.")

    recent_seq = recent.iloc[-seq_len:].copy()
    # fill any missing technicals
    recent_seq = add_features(recent_seq).fillna(method='ffill').fillna(method='bfill').replace([np.inf,-np.inf],0)

    # create scaled feature vector using previously fitted scaler
    feat_vals = recent_seq[features].values
    scaled_vals = scaler.transform(recent_seq[features])  # shape (seq_len, n_features)
    x = torch.tensor(scaled_vals[np.newaxis,:,:], dtype=torch.float32).to(DEVICE)  # 1 x seq_len x features

    # model predictions
    preds = []
    for m in models:
        m.eval()
        with torch.no_grad():
            out = m(x).squeeze(-1).cpu().numpy().ravel()
            preds.append(float(out[0]))
    # average return (these are returns in scaled-space: careful — Target earlier was raw return not scaled)
    # In our pipeline 'Target' was not scaled; however we trained models on scaled 'Target' if we used fit_transform on target.
    # Here we constructed target in scaled form when building dataset earlier (we used columns from scaled df),
    # so the model outputs are on the scaled TARGET scale. We must invert scaling for return.
    # To invert, we will reconstruct an array with one sample and use scaler inverse on the feature that corresponds to 'Target'.
    # Simpler approach: since we trained on scaled Target (target was part of scaled frame earlier),
    # but MinMaxScaler was fit only on features (we didn't scale Target explicitly). In our dataset we used df_scaled[target]
    # by using the scaled frame for target; to keep consistent we will assume the models directly predict raw return values.
    # For safety: we'll treat outputs as raw returns (as in original code). If your training used a different target scaling, adjust here.

    avg_pred_return = np.mean(preds)

    # compute predicted next price using last actual close
    last_close = recent_seq['Close'].values[-1]
    predicted_price = float(last_close * (1 + avg_pred_return))

    return {'next_trade_date': next_trade, 'predicted_return': float(avg_pred_return), 'predicted_price': float(predicted_price), 'component_returns': preds}

# Run next-day prediction
ensemble_models = [tlstm_model.to(DEVICE), gru_model.to(DEVICE), trans_model.to(DEVICE)]
try:
    next_pred = predict_next_day(ensemble_models, scaler, stock_df, features, SEQ_LEN)
    print("\n--- NEXT TRADING DAY PREDICTION (Realtime) ---")
    print(f"Next trading day (calendar): {next_pred['next_trade_date']}")
    print(f"Predicted return (next day): {next_pred['predicted_return']:.6f}")
    print(f"Predicted Close price (next day): ₹{next_pred['predicted_price']:.2f}")
    print(f"Ensemble component raw preds: {next_pred['component_returns']}")
except Exception as e:
    print("Next day prediction failed:", e)

# ---------------- Optional: plot last test segment vs predictions ----------------
if len(tlstm_preds) and len(final_predicted_returns):
    plt.figure(figsize=(12,6))
    plt.plot(test_df_unscaled['Date'][SEQ_LEN:], actual_prices, label='Actual Price')
    plt.plot(test_df_unscaled['Date'][SEQ_LEN:], predicted_prices, label='Predicted Price (Ensemble)', linestyle='--')
    plt.title(f"{TICKER} - Actual vs Predicted (Ensemble)")
    plt.xlabel("Date")
    plt.ylabel("Close Price (INR)")
    plt.legend()
    plt.tight_layout()
    plt.show()

# ------------- Notes & usage -------------
# - To make this truly "live" in production:
#   * run this cell right after the market close or whenever you want the next-day estimate
#   * you can schedule this cell or script using cron/Cloud Run; for Colab manual run gives "real-time" next-day post-training.
# - If you want to persist models and reuse without re-training each run, save model.state_dict() after training and load for prediction only.
# - If you enable USE_FINBERT=True, download time is higher but sentiment quality for finance may improve.
# - If you want to predict a probability or confidence band, compute ensemble variance and/or bootstrapped runs.
```
