```python
Install necessary libraries in the Colab environment
!pip install yfinance transformers torch scikit-learn requests tqdm

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from datetime import datetime, timedelta
from tqdm import tqdm
import requests
import yfinance as yf
import warnings

warnings.filterwarnings('ignore')

# --- Constants ---
SEQ_LEN = 30
BATCH_SIZE = 32
EPOCHS = 30 # Increased epochs slightly to compensate for lower learning rate
# FIX: Lowered learning rate for more stable training
LEARNING_RATE = 0.0001
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
NEWS_API_KEY = '99d1937942dd430cb7e67bbd39c0a713'

print(f"Using device: {DEVICE}")

# --- Data Fetching & Cleaning ---
def fetch_stock_data(ticker="AAPL"):
    """Fetches daily stock data and cleans it to prevent merge errors."""
    df = yf.download(tickers=ticker, period="5y", interval="1d")

    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df.reset_index(inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    return df

def fetch_news(symbol="Apple"):
    """Fetches news from the last 30 days."""
    to_date = datetime.today().strftime('%Y-%m-%d')
    from_date = (datetime.today() - timedelta(days=30)).strftime('%Y-%m-%d')
    url = f"https://newsapi.org/v2/everything?q={symbol}&from={from_date}&to={to_date}&language=en&sortBy=publishedAt&apiKey={NEWS_API_KEY}"
    try:
        res = requests.get(url).json()
        if res.get("status") == "ok":
            articles = res.get("articles", [])
            headlines = [{"Date": a['publishedAt'][:10], "Headline": a['title']} for a in articles]
            return pd.DataFrame(headlines)
        else:
            return pd.DataFrame(columns=["Date", "Headline"])
    except Exception:
        return pd.DataFrame(columns=["Date", "Headline"])

# --- Dummy Events ---
events_df = pd.DataFrame({
    "Date": [datetime.today() - timedelta(days=5), datetime.today() - timedelta(days=2)],
    "Event_Type": ["Product Launch", "Fed Announcement"]
})
events_df['Date'] = pd.to_datetime(events_df['Date'])

# --- Load Data ---
stock_df = fetch_stock_data("AAPL")
news_df = fetch_news("Apple")

# --- FinBERT Sentiment Scoring ---
if not news_df.empty:
    tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
    model_fbert = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone").to(DEVICE)

    def get_sentiment(text):
        if not text or not isinstance(text, str): return 0.0
        inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True).to(DEVICE)
        outputs = model_fbert(**inputs)
        scores = torch.nn.functional.softmax(outputs.logits, dim=1).detach().cpu().numpy()[0]
        return scores[1] - scores[0]

    news_df['Sentiment'] = news_df['Headline'].apply(get_sentiment)
    sentiment_df = news_df.groupby("Date")["Sentiment"].mean().reset_index()
    sentiment_df['Date'] = pd.to_datetime(sentiment_df['Date'])
else:
    sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])

# --- Feature Engineering ---
def add_features(df):
    df['Return'] = df['Close'].pct_change()
    df['MA10'] = df['Close'].rolling(window=10).mean()
    df['MA50'] = df['Close'].rolling(window=50).mean()
    delta = df['Close'].diff()
    up, down = delta.clip(lower=0), -1*delta.clip(upper=0)
    ema_up, ema_down = up.ewm(com=13, adjust=False).mean(), down.ewm(com=13, adjust=False).mean()
    rs = ema_up / ema_down
    df['RSI'] = 100 - (100 / (1 + rs))
    df['MACD'] = df['Close'].ewm(span=12, adjust=False).mean() - df['Close'].ewm(span=26, adjust=False).mean()
    df['Volatility'] = df['Return'].rolling(window=21).std() * np.sqrt(252)
    return df

stock_df = add_features(stock_df)

# --- Merge Data ---
events_df['Date'] = pd.to_datetime(events_df['Date'])
event_dummies = pd.get_dummies(events_df['Event_Type'], prefix='Event', dtype=float)
events_df = pd.concat([events_df['Date'], event_dummies], axis=1)

stock_df = pd.merge(stock_df, events_df, on='Date', how='left')
if not sentiment_df.empty:
    stock_df = pd.merge(stock_df, sentiment_df, on='Date', how='left')
else:
    stock_df['Sentiment'] = 0.0

# --- FIX: Change the target variable to be the next day's return ---
stock_df['Target'] = stock_df['Return'].shift(-1)
stock_df.fillna(0, inplace=True)
stock_df.replace([np.inf, -np.inf], 0, inplace=True) # Handle potential inf values from RSI calc
stock_df.dropna(inplace=True) # Drop last row with no target

# --- Normalize & Split Data ---
event_cols = [col for col in stock_df.columns if 'Event_' in col]
features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA10', 'MA50', 'RSI', 'MACD', 'Return', 'Volatility', 'Sentiment'] + event_cols

scaler = MinMaxScaler()
stock_df[features] = scaler.fit_transform(stock_df[features])

train_size = int(len(stock_df) * 0.85)
train_df = stock_df.iloc[:train_size].copy()
test_df = stock_df.iloc[train_size:].copy()

# --- PyTorch Dataset ---
class StockDataset(Dataset):
    def __init__(self, df, features, target_col, seq_len):
        self.X, self.y = [], []
        # Store original close prices from the test set for later evaluation
        self.close_prices_for_eval = df['Close'].values[seq_len:]

        feature_data = df[features].values
        target_data = df[target_col].values

        for i in range(seq_len, len(df)):
            self.X.append(feature_data[i-seq_len:i])
            self.y.append(target_data[i])

        self.X = np.array(self.X)
        self.y = np.array(self.y)

    def __len__(self): return len(self.X)
    def __getitem__(self, idx):
        return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)

train_dataset = StockDataset(train_df, features, 'Target', SEQ_LEN)
test_dataset = StockDataset(test_df, features, 'Target', SEQ_LEN)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

# --- Model Definitions (remain the same) ---
class TLSTM_Attention(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(TLSTM_Attention, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.attention = nn.Linear(hidden_size, 1)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
        context = torch.sum(attn_weights * lstm_out, dim=1)
        return self.fc(context)

class GRU_Model(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(GRU_Model, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.gru(x)
        return self.fc(out[:, -1, :])

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.attn = nn.MultiheadAttention(embed_dim, num_heads, batch_first=True, dropout=rate)
        self.ffn = nn.Sequential(nn.Linear(embed_dim, ff_dim), nn.ReLU(), nn.Linear(ff_dim, embed_dim))
        self.layernorm1, self.layernorm2 = nn.LayerNorm(embed_dim, eps=1e-6), nn.LayerNorm(embed_dim, eps=1e-6)
        self.dropout1, self.dropout2 = nn.Dropout(rate), nn.Dropout(rate)
        self.fc_out = nn.Linear(embed_dim, 1)
    def forward(self, x):
        attn_output, _ = self.attn(x, x, x)
        out1 = self.layernorm1(x + self.dropout1(attn_output))
        ffn_output = self.ffn(out1)
        out2 = self.layernorm2(out1 + self.dropout2(ffn_output))
        return self.fc_out(out2.mean(dim=1))

# --- Training and Evaluation ---
def train_model(model, train_loader, epochs, lr):
    loss_fn = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.train()
    print(f"\n--- Training {model.__class__.__name__} ---")
    for epoch in range(epochs):
        total_loss = 0
        for X_batch, y_batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
            optimizer.zero_grad()
            preds = model(X_batch).squeeze(-1)
            loss = loss_fn(preds, y_batch)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch+1}/{epochs} - Loss: {total_loss / len(train_loader):.8f}")
    return model

def get_predictions(model, loader):
    model.eval()
    preds = []
    with torch.no_grad():
        for X_batch, _ in loader:
            preds.append(model(X_batch.to(DEVICE)).squeeze(-1).cpu().numpy())
    return np.concatenate(preds)

# --- Instantiate and Train All Models ---
input_size = len(features)
hidden_size = 64

tlstm_model = train_model(TLSTM_Attention(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
gru_model = train_model(GRU_Model(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
trans_model = train_model(TransformerBlock(input_size, 2, 64).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)

# --- Ensemble Prediction ---
tlstm_preds = get_predictions(tlstm_model, test_loader)
gru_preds = get_predictions(gru_model, test_loader)
trans_preds = get_predictions(trans_model, test_loader)

# FIX: Simplify the ensemble to a simple average for robustness
final_predicted_returns = (tlstm_preds + gru_preds + trans_preds) / 3.0

# --- Convert Predicted Returns back to Prices for Evaluation ---
# Get the actual close prices from the day before the prediction
last_actual_prices = test_dataset.close_prices_for_eval[:-1]
# Calculate the predicted prices
predicted_prices = last_actual_prices * (1 + final_predicted_returns[:-1])
# Get the true actual prices for comparison
actual_prices = test_dataset.close_prices_for_eval[1:]

# --- Evaluation ---
print("\n--- Final Ensemble Model Evaluation ---")
mae = mean_absolute_error(actual_prices, predicted_prices)
mse = mean_squared_error(actual_prices, predicted_prices)
r2 = r2_score(actual_prices, predicted_prices)
print(f"Mean Absolute Error (MAE): ${mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R²): {r2:.4f}")

# --- Plot Results ---
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(15, 7))
plt.plot(actual_prices, label='Actual Price', color='royalblue', linewidth=2)
plt.plot(predicted_prices, label='Predicted Price', color='crimson', linestyle='--')
plt.title("AAPL Stock Price Prediction: Actual vs. Predicted (Ensemble)", fontsize=16)
plt.xlabel("Time (Days in Test Set)", fontsize=12)
plt.ylabel("Unscaled Close Price (USD)", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# NEW: Install necessary libraries in the Colab environment
!pip install yfinance transformers torch scikit-learn requests tqdm

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from datetime import datetime, timedelta
from tqdm import tqdm
import requests
import yfinance as yf
import warnings
import re

warnings.filterwarnings('ignore')

# --- Constants ---
SEQ_LEN = 30
BATCH_SIZE = 32
EPOCHS = 100
LEARNING_RATE = 0.0001
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
# --- IMPORTANT: Replace with your own NewsAPI key ---
NEWS_API_KEY = '99d1937942dd430cb7e67bbd39c0a713' # <--- REPLACE THIS

print(f"Using device: {DEVICE}")

# --- Data Fetching & Cleaning ---
def fetch_stock_data(ticker):
    """Fetches daily stock data for the last 5 years and cleans it."""
    df = yf.download(tickers=ticker, period="5y", interval="1d", progress=False)
    if df.empty: return None
    if isinstance(df.columns, pd.MultiIndex): df.columns = df.columns.get_level_values(0)
    df.reset_index(inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    return df

def fetch_news(symbol):
    """Fetches news from the last 30 days from NewsAPI."""
    if NEWS_API_KEY == 'YOUR_NEWS_API_KEY':
        print("Warning: NewsAPI key not provided. Skipping news fetching.")
        return pd.DataFrame(columns=["Date", "Headline"])

    to_date = datetime.today().strftime('%Y-%m-%d')
    from_date = (datetime.today() - timedelta(days=30)).strftime('%Y-%m-%d')
    url = f"https://newsapi.org/v2/everything?q={symbol}&from={from_date}&to={to_date}&language=en&sortBy=publishedAt&apiKey={NEWS_API_KEY}"
    try:
        res = requests.get(url).json()
        if res.get("status") == "ok":
            articles = res.get("articles", [])
            headlines = [{"Date": a['publishedAt'][:10], "Headline": a['title']} for a in articles]
            news_df = pd.DataFrame(headlines)
            news_df['Date'] = pd.to_datetime(news_df['Date'])
            return news_df
        else:
            print(f"Error fetching news: {res.get('message')}")
            return pd.DataFrame(columns=["Date", "Headline"])
    except Exception as e:
        print(f"An exception occurred during news fetching: {e}")
        return pd.DataFrame(columns=["Date", "Headline"])

# MODIFIED: Function to automatically find company-specific and market events
def fetch_real_events(ticker_symbol, stock_df):
    """
    Fetches real financial events from yfinance (earnings, dividends, splits)
    and includes general market events (RBI meetings).
    """
    all_events = []
    ticker_obj = yf.Ticker(ticker_symbol)

    # 1. Fetch Earnings Dates
    try:
        earnings_dates = ticker_obj.get_earnings_dates(limit=20)
        if earnings_dates is not None and not earnings_dates.empty:
            earnings_df = pd.DataFrame({
                'Date': pd.to_datetime(earnings_dates.index.date),
                'Event_Type': 'Earnings'
            })
            all_events.append(earnings_df)
    except Exception as e:
        print(f"Could not fetch earnings dates: {e}")

    # 2. NEW: Fetch Dividends and Stock Splits
    try:
        actions_df = ticker_obj.actions
        if actions_df is not None and not actions_df.empty:
            actions_df.reset_index(inplace=True)
            actions_df['Date'] = pd.to_datetime(actions_df['Date'].dt.date)

            dividends = actions_df[actions_df['Dividends'] > 0][['Date']].copy()
            dividends['Event_Type'] = 'Dividend'

            splits = actions_df[actions_df['Stock Splits'] > 0][['Date']].copy()
            splits['Event_Type'] = 'Stock Split'

            if not dividends.empty: all_events.append(dividends)
            if not splits.empty: all_events.append(splits)
    except Exception as e:
        print(f"Could not fetch corporate actions (dividends/splits): {e}")

    # 3. Add RBI MPC Meeting Dates (general market event)
    rbi_mpc_dates = [
        '2020-02-06', '2020-03-27', '2020-05-22', '2020-08-06', '2020-10-09', '2020-12-04', '2021-02-05', '2021-04-07', '2021-06-04', '2021-08-06', '2021-10-08', '2021-12-08', '2022-02-10', '2022-04-08', '2022-05-04', '2022-06-08', '2022-08-05', '2022-09-30', '2022-12-07', '2023-02-08', '2023-04-06', '2023-06-08', '2023-08-10', '2023-10-06', '2023-12-08', '2024-02-08', '2024-04-05', '2024-06-07', '2024-08-08', '2024-10-09', '2024-12-06', '2025-02-07', '2025-04-04', '2025-06-06', '2025-08-08'
    ]
    rbi_df = pd.DataFrame({'Date': pd.to_datetime(rbi_mpc_dates), 'Event_Type': 'RBI MPC Meeting'})
    all_events.append(rbi_df)

    if not all_events: return pd.DataFrame(columns=['Date'])

    events_df = pd.concat(all_events, ignore_index=True)
    min_date, max_date = stock_df['Date'].min(), stock_df['Date'].max()
    events_df = events_df[(events_df['Date'] >= min_date) & (events_df['Date'] <= max_date)]
    return events_df.drop_duplicates(subset=['Date', 'Event_Type']).reset_index(drop=True)

# NEW: Function to find major events by scanning news headlines for keywords
def find_events_from_news(news_df):
    """Scans news headlines for keywords to identify major announcements."""
    keyword_map = {
        'Major Announcement': [r'AGM', r'annual general meeting', r'shareholder meeting', r'board meeting'],
        'Product Launch': [r'launch', r'unveil', r'release', r'introduce'],
        'Merger/Acquisition': [r'merger', r'acquire', r'acquisition', r'takeover', r'stake'],
        'Partnership/Deal': [r'partnership', r'deal', r'agreement', r'collaboration', r'joint venture']
    }
    found_events = []
    if news_df.empty: return pd.DataFrame(columns=['Date', 'Event_Type'])

    for _, row in news_df.iterrows():
        headline = row['Headline'].lower()
        for event_type, keywords in keyword_map.items():
            if any(re.search(keyword, headline) for keyword in keywords):
                found_events.append({'Date': row['Date'], 'Event_Type': event_type})
                break # Classify each headline only once

    if not found_events: return pd.DataFrame(columns=['Date', 'Event_Type'])
    return pd.DataFrame(found_events).drop_duplicates()

# --- Feature Engineering & Preprocessing (remains the same) ---
def add_features(df):
    df['Return'] = df['Close'].pct_change()
    df['MA10'] = df['Close'].rolling(window=10).mean()
    df['MA50'] = df['Close'].rolling(window=50).mean()
    delta = df['Close'].diff()
    up, down = delta.clip(lower=0), -1*delta.clip(upper=0)
    ema_up, ema_down = up.ewm(com=13, adjust=False).mean(), down.ewm(com=13, adjust=False).mean()
    rs = ema_up / ema_down
    df['RSI'] = 100 - (100 / (1 + rs))
    df['MACD'] = df['Close'].ewm(span=12, adjust=False).mean() - df['Close'].ewm(span=26, adjust=False).mean()
    df['Volatility'] = df['Return'].rolling(window=21).std() * np.sqrt(252)
    return df

# --- Model Definitions (remain the same) ---
class TLSTM_Attention(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(TLSTM_Attention, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.attention = nn.Linear(hidden_size, 1)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
        context = torch.sum(attn_weights * lstm_out, dim=1)
        return self.fc(context)

class GRU_Model(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(GRU_Model, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.gru(x)
        return self.fc(out[:, -1, :])

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.attn = nn.MultiheadAttention(embed_dim, num_heads, batch_first=True, dropout=rate)
        self.ffn = nn.Sequential(nn.Linear(embed_dim, ff_dim), nn.ReLU(), nn.Linear(ff_dim, embed_dim))
        self.layernorm1, self.layernorm2 = nn.LayerNorm(embed_dim, eps=1e-6), nn.LayerNorm(embed_dim, eps=1e-6)
        self.dropout1, self.dropout2 = nn.Dropout(rate), nn.Dropout(rate)
        self.fc_out = nn.Linear(embed_dim, 1)
    def forward(self, x):
        attn_output, _ = self.attn(x, x, x)
        out1 = self.layernorm1(x + self.dropout1(attn_output))
        ffn_output = self.ffn(out1)
        out2 = self.layernorm2(out1 + self.dropout2(ffn_output))
        return self.fc_out(out2.mean(dim=1))

# --- Training and Evaluation Functions (remain the same) ---
def train_model(model, train_loader, epochs, lr):
    loss_fn = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.train()
    print(f"\n--- Training {model.__class__.__name__} ---")
    for epoch in range(epochs):
        total_loss = 0
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")
        for X_batch, y_batch in pbar:
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
            optimizer.zero_grad()
            preds = model(X_batch).squeeze(-1)
            loss = loss_fn(preds, y_batch)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
            pbar.set_postfix({'Loss': f"{total_loss / len(pbar):.8f}"})
    return model

def get_predictions(model, loader):
    model.eval()
    preds = []
    with torch.no_grad():
        for X_batch, _ in loader:
            preds.append(model(X_batch.to(DEVICE)).squeeze(-1).cpu().numpy())
    return np.concatenate(preds)

# --- PyTorch Dataset (remains the same) ---
class StockDataset(Dataset):
    def __init__(self, df, df_unscaled, features, target_col, seq_len):
        self.X, self.y, self.close_prices_for_eval = [], [], df_unscaled['Close'].values[seq_len-1:-1]
        feature_data, target_data = df[features].values, df[target_col].values
        for i in range(seq_len, len(df)):
            self.X.append(feature_data[i-seq_len:i])
            self.y.append(target_data[i])
        self.X, self.y = np.array(self.X), np.array(self.y)
    def __len__(self): return len(self.X)
    def __getitem__(self, idx): return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)

# --- MODIFIED: Main function to run the interactive process ---
def main():
    # 1. Get and Validate User Input
    while True:
        COMPANY_NAME = input("▶️ Enter the full company name for news search (e.g., Tata Motors): ")
        TICKER = input(f"▶️ Enter the Yahoo Finance ticker for '{COMPANY_NAME}' (e.g., TATAMOTORS.NS): ")

        print(f"\nValidating ticker '{TICKER}'...")
        stock_df = fetch_stock_data(TICKER)

        if stock_df is not None:
            print(f"✅ Ticker '{TICKER}' is valid. Starting analysis...")
            break
        else:
            print(f"❌ Error: Ticker '{TICKER}' is invalid or no data was found. Please try again.")
            print("---")

    # 2. Fetch Data
    print("\n--- Fetching Data ---")
    news_df = fetch_news(COMPANY_NAME)
    # MODIFIED: Fetch events from both financial data and news keywords
    financial_events_df = fetch_real_events(TICKER, stock_df)
    news_events_df = find_events_from_news(news_df)

    # MODIFIED: Combine all found events
    events_df = pd.concat([financial_events_df, news_events_df], ignore_index=True)
    print(f"Found {len(events_df)} potential event entries.")


    # 3. Sentiment Scoring
    if not news_df.empty:
        print("\n--- Calculating News Sentiment ---")
        try:
            tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
            model_fbert = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone").to(DEVICE)
            def get_sentiment(text):
                if not text or not isinstance(text, str): return 0.0
                inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True).to(DEVICE)
                with torch.no_grad(): outputs = model_fbert(**inputs)
                scores = torch.nn.functional.softmax(outputs.logits, dim=1).cpu().numpy()[0]
                return scores[1] - scores[0]
            tqdm.pandas(desc="Analyzing headlines")
            news_df['Sentiment'] = news_df['Headline'].progress_apply(get_sentiment)
            sentiment_df = news_df.groupby("Date")["Sentiment"].mean().reset_index()
            sentiment_df['Date'] = pd.to_datetime(sentiment_df['Date'])
        except Exception as e:
            print(f"Error during sentiment analysis: {e}")
            sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])
    else:
        sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])

    # 4. Feature Engineering and Merging
    print("\n--- Engineering Features ---")
    stock_df = add_features(stock_df)
    if not events_df.empty:
        event_dummies = pd.get_dummies(events_df['Event_Type'], prefix='Event', dtype=float)
        events_df = pd.concat([events_df['Date'], event_dummies], axis=1).groupby('Date').sum().reset_index()
        stock_df = pd.merge(stock_df, events_df, on='Date', how='left')
    if not sentiment_df.empty:
        stock_df = pd.merge(stock_df, sentiment_df, on='Date', how='left')
    else:
        stock_df['Sentiment'] = 0.0

    event_cols = [col for col in stock_df.columns if 'Event_' in col]
    stock_df['Target'] = stock_df['Return'].shift(-1)
    stock_df[event_cols] = stock_df[event_cols].fillna(0)
    stock_df['Sentiment'].fillna(0, inplace=True)
    tech_cols = ['Return', 'MA10', 'MA50', 'RSI', 'MACD', 'Volatility']
    stock_df[tech_cols] = stock_df[tech_cols].fillna(method='ffill').fillna(method='bfill')
    stock_df.replace([np.inf, -np.inf], 0, inplace=True)
    stock_df.dropna(inplace=True)

    # 5. Normalize & Split Data
    features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA10', 'MA50', 'RSI', 'MACD', 'Return', 'Volatility', 'Sentiment'] + event_cols
    scaler = MinMaxScaler()
    stock_df_scaled = stock_df.copy()
    stock_df_scaled[features] = scaler.fit_transform(stock_df_scaled[features])
    train_size = int(len(stock_df_scaled) * 0.85)
    train_df, test_df = stock_df_scaled.iloc[:train_size].copy(), stock_df_scaled.iloc[train_size:].copy()
    test_df_unscaled = stock_df.iloc[train_size:].copy()

    # 6. Create Datasets and Dataloaders
    train_dataset = StockDataset(train_df, train_df, features, 'Target', SEQ_LEN)
    test_dataset = StockDataset(test_df, test_df_unscaled, features, 'Target', SEQ_LEN)
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

    # 7. Train Models
    input_size = len(features)
    hidden_size = 64
    num_heads = 3 if input_size % 3 == 0 else (2 if input_size % 2 == 0 else 1)
    print(f"\nTotal input features: {input_size}. Using {num_heads} attention heads for Transformer.")

    tlstm_model = train_model(TLSTM_Attention(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
    gru_model = train_model(GRU_Model(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
    trans_model = train_model(TransformerBlock(input_size, num_heads, 64).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)

    # 8. Ensemble Predictions and Evaluate
    tlstm_preds, gru_preds, trans_preds = get_predictions(tlstm_model, test_loader), get_predictions(gru_model, test_loader), get_predictions(trans_model, test_loader)
    final_predicted_returns = (tlstm_preds + gru_preds + trans_preds) / 3.0
    last_actual_prices = test_dataset.close_prices_for_eval
    predicted_prices = last_actual_prices * (1 + final_predicted_returns)
    actual_prices = test_df_unscaled['Close'].values[SEQ_LEN:]

    print("\n--- Final Ensemble Model Evaluation ---")
    mae, mse, r2 = mean_absolute_error(actual_prices, predicted_prices), mean_squared_error(actual_prices, predicted_prices), r2_score(actual_prices, predicted_prices)
    print(f"Mean Absolute Error (MAE): ₹{mae:.2f}")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"R-squared (R²): {r2:.4f}")

    # 9. Plot Results
    plt.style.use('seaborn-v0_8-darkgrid')
    plt.figure(figsize=(15, 7))
    plt.plot(test_df_unscaled['Date'][SEQ_LEN:], actual_prices, label='Actual Price', color='royalblue', linewidth=2)
    plt.plot(test_df_unscaled['Date'][SEQ_LEN:], predicted_prices, label='Predicted Price', color='crimson', linestyle='--')
    plt.title(f"{TICKER} Stock Price Prediction: Actual vs. Predicted (Ensemble)", fontsize=16)
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Close Price (INR)", fontsize=12)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Run the main function when the script is executed
if __name__ == "__main__":
    main()
```

```python
# NEW: Install necessary libraries, including Plotly for candlestick charts
!pip install yfinance transformers torch scikit-learn requests tqdm plotly

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import plotly.graph_objects as go # NEW: Import Plotly
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from datetime import datetime, timedelta
from tqdm import tqdm
import requests
import yfinance as yf
import warnings
import re

warnings.filterwarnings('ignore')

# --- Constants ---
SEQ_LEN = 30
BATCH_SIZE = 32
EPOCHS = 100
LEARNING_RATE = 0.0001
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
# --- IMPORTANT: Replace with your own NewsAPI key ---
NEWS_API_KEY = '99d1937942dd430cb7e67bbd39c0a713' # <--- REPLACE THIS

print(f"Using device: {DEVICE}")

# --- Data Fetching & Cleaning (remains the same) ---
def fetch_stock_data(ticker):
    df = yf.download(tickers=ticker, period="5y", interval="1d", progress=False)
    if df.empty: return None
    if isinstance(df.columns, pd.MultiIndex): df.columns = df.columns.get_level_values(0)
    df.reset_index(inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    return df

def fetch_news(symbol):
    if NEWS_API_KEY == 'YOUR_NEWS_API_KEY':
        return pd.DataFrame(columns=["Date", "Headline"])
    to_date = datetime.today().strftime('%Y-%m-%d')
    from_date = (datetime.today() - timedelta(days=30)).strftime('%Y-%m-%d')
    url = f"https://newsapi.org/v2/everything?q={symbol}&from={from_date}&to={to_date}&language=en&sortBy=publishedAt&apiKey={NEWS_API_KEY}"
    try:
        res = requests.get(url).json()
        if res.get("status") == "ok":
            articles = res.get("articles", [])
            headlines = [{"Date": a['publishedAt'][:10], "Headline": a['title']} for a in articles]
            news_df = pd.DataFrame(headlines)
            news_df['Date'] = pd.to_datetime(news_df['Date'])
            return news_df
        else:
            return pd.DataFrame(columns=["Date", "Headline"])
    except Exception as e:
        return pd.DataFrame(columns=["Date", "Headline"])

def fetch_real_events(ticker_symbol, stock_df):
    all_events, ticker_obj = [], yf.Ticker(ticker_symbol)
    try: # Earnings
        earnings_dates = ticker_obj.get_earnings_dates(limit=20)
        if earnings_dates is not None and not earnings_dates.empty:
            all_events.append(pd.DataFrame({'Date': pd.to_datetime(earnings_dates.index.date), 'Event_Type': 'Earnings'}))
    except Exception: pass
    try: # Dividends & Splits
        actions_df = ticker_obj.actions
        if actions_df is not None and not actions_df.empty:
            actions_df.reset_index(inplace=True)
            actions_df['Date'] = pd.to_datetime(actions_df['Date'].dt.date)
            dividends = actions_df[actions_df['Dividends'] > 0][['Date']].copy()
            dividends['Event_Type'] = 'Dividend'
            splits = actions_df[actions_df['Stock Splits'] > 0][['Date']].copy()
            splits['Event_Type'] = 'Stock Split'
            if not dividends.empty: all_events.append(dividends)
            if not splits.empty: all_events.append(splits)
    except Exception: pass
    rbi_mpc_dates = ['2020-02-06', '2020-03-27', '2020-05-22', '2020-08-06', '2020-10-09', '2020-12-04', '2021-02-05', '2021-04-07', '2021-06-04', '2021-08-06', '2021-10-08', '2021-12-08', '2022-02-10', '2022-04-08', '2022-05-04', '2022-06-08', '2022-08-05', '2022-09-30', '2022-12-07', '2023-02-08', '2023-04-06', '2023-06-08', '2023-08-10', '2023-10-06', '2023-12-08', '2024-02-08', '2024-04-05', '2024-06-07', '2024-08-08', '2024-10-09', '2024-12-06', '2025-02-07', '2025-04-04', '2025-06-06', '2025-08-08']
    all_events.append(pd.DataFrame({'Date': pd.to_datetime(rbi_mpc_dates), 'Event_Type': 'RBI MPC Meeting'}))
    if not all_events: return pd.DataFrame(columns=['Date'])
    events_df = pd.concat(all_events, ignore_index=True)
    min_date, max_date = stock_df['Date'].min(), stock_df['Date'].max()
    events_df = events_df[(events_df['Date'] >= min_date) & (events_df['Date'] <= max_date)]
    return events_df.drop_duplicates(subset=['Date', 'Event_Type']).reset_index(drop=True)

def find_events_from_news(news_df):
    keyword_map = {'Major Announcement': [r'AGM', r'annual general meeting'], 'Product Launch': [r'launch', r'unveil'], 'Merger/Acquisition': [r'merger', r'acquire', r'takeover'], 'Partnership/Deal': [r'partnership', r'deal']}
    found_events = []
    if news_df.empty: return pd.DataFrame(columns=['Date', 'Event_Type'])
    for _, row in news_df.iterrows():
        headline = row['Headline'].lower()
        for event_type, keywords in keyword_map.items():
            if any(re.search(keyword, headline) for keyword in keywords):
                found_events.append({'Date': row['Date'], 'Event_Type': event_type}); break
    if not found_events: return pd.DataFrame(columns=['Date', 'Event_Type'])
    return pd.DataFrame(found_events).drop_duplicates()

# --- Feature Engineering & Preprocessing ---
# MODIFIED: Added MACD Signal Line
def add_features(df):
    df['Return'] = df['Close'].pct_change()
    df['MA10'] = df['Close'].rolling(window=10).mean()
    df['MA50'] = df['Close'].rolling(window=50).mean()
    delta = df['Close'].diff()
    up, down = delta.clip(lower=0), -1*delta.clip(upper=0)
    ema_up, ema_down = up.ewm(com=13, adjust=False).mean(), down.ewm(com=13, adjust=False).mean()
    rs = ema_up / ema_down
    df['RSI'] = 100 - (100 / (1 + rs))
    df['MACD'] = df['Close'].ewm(span=12, adjust=False).mean() - df['Close'].ewm(span=26, adjust=False).mean()
    df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean() # NEW: Calculate signal line
    df['Volatility'] = df['Return'].rolling(window=21).std() * np.sqrt(252)
    return df

# --- Model Definitions (remain the same) ---
class TLSTM_Attention(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(TLSTM_Attention, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.attention = nn.Linear(hidden_size, 1); self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        lstm_out, _ = self.lstm(x); attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
        context = torch.sum(attn_weights * lstm_out, dim=1); return self.fc(context)

class GRU_Model(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(GRU_Model, self).__init__(); self.gru = nn.GRU(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.gru(x); return self.fc(out[:, -1, :])

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.attn = nn.MultiheadAttention(embed_dim, num_heads, batch_first=True, dropout=rate)
        self.ffn = nn.Sequential(nn.Linear(embed_dim, ff_dim), nn.ReLU(), nn.Linear(ff_dim, embed_dim))
        self.layernorm1, self.layernorm2 = nn.LayerNorm(embed_dim, eps=1e-6), nn.LayerNorm(embed_dim, eps=1e-6)
        self.dropout1, self.dropout2 = nn.Dropout(rate), nn.Dropout(rate); self.fc_out = nn.Linear(embed_dim, 1)
    def forward(self, x):
        attn_output, _ = self.attn(x, x, x); out1 = self.layernorm1(x + self.dropout1(attn_output))
        ffn_output = self.ffn(out1); out2 = self.layernorm2(out1 + self.dropout2(ffn_output))
        return self.fc_out(out2.mean(dim=1))

# --- Training, Evaluation, Dataset (remain the same) ---
def train_model(model, train_loader, epochs, lr):
    loss_fn = nn.MSELoss(); optimizer = optim.Adam(model.parameters(), lr=lr); model.train()
    print(f"\n--- Training {model.__class__.__name__} ---")
    for epoch in range(epochs):
        total_loss = 0; pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")
        for X_batch, y_batch in pbar:
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE); optimizer.zero_grad()
            preds = model(X_batch).squeeze(-1); loss = loss_fn(preds, y_batch); loss.backward(); optimizer.step()
            total_loss += loss.item(); pbar.set_postfix({'Loss': f"{total_loss / len(pbar):.8f}"})
    return model

def get_predictions(model, loader):
    model.eval(); preds = []
    with torch.no_grad():
        for X_batch, _ in loader: preds.append(model(X_batch.to(DEVICE)).squeeze(-1).cpu().numpy())
    return np.concatenate(preds)

class StockDataset(Dataset):
    def __init__(self, df, df_unscaled, features, target_col, seq_len):
        self.X, self.y, self.close_prices_for_eval = [], [], df_unscaled['Close'].values[seq_len-1:-1]
        feature_data, target_data = df[features].values, df[target_col].values
        for i in range(seq_len, len(df)): self.X.append(feature_data[i-seq_len:i]); self.y.append(target_data[i])
        self.X, self.y = np.array(self.X), np.array(self.y)
    def __len__(self): return len(self.X)
    def __getitem__(self, idx): return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)

# --- NEW: Signal Generation and Candlestick Plotting Functions ---
def generate_trading_signal(last_day_data, predicted_next_return):
    """Generates a buy/don't buy signal based on a set of rules."""
    signal_checks = {
        "Model Prediction is Positive": predicted_next_return > 0,
        "Uptrend (MA10 > MA50)": last_day_data['MA10'] > last_day_data['MA50'],
        "Momentum is Bullish (MACD > Signal)": last_day_data['MACD'] > last_day_data['MACD_Signal'],
        "Not Overbought (RSI < 75)": last_day_data['RSI'] < 75
    }

    print("\n--- Trading Signal Analysis ---")
    print("Disclaimer: This is not financial advice. For educational purposes only.")
    print(f"Analysis for date: {last_day_data['Date'].strftime('%Y-%m-%d')}")
    print("-" * 30)

    all_conditions_met = True
    for check, result in signal_checks.items():
        print(f"▶️ {check}: {'✅' if result else '❌'}")
        if not result:
            all_conditions_met = False

    print("-" * 30)
    if all_conditions_met:
        print("Signal: 🟢 BUY 🟢")
        print("Reasoning: All bullish conditions were met, including a positive model prediction.")
    else:
        print("Signal: 🔴 HOLD / DO NOT BUY 🔴")
        print("Reasoning: One or more conditions for a buy signal were not met.")

def plot_candlestick_with_signals(df, ticker):
    """Plots an interactive candlestick chart for the last 120 days."""
    recent_df = df.tail(120)

    fig = go.Figure()
    # Candlestick chart
    fig.add_trace(go.Candlestick(x=recent_df['Date'],
                    open=recent_df['Open'], high=recent_df['High'],
                    low=recent_df['Low'], close=recent_df['Close'],
                    name='Price'))
    # Moving Averages
    fig.add_trace(go.Scatter(x=recent_df['Date'], y=recent_df['MA10'], line=dict(color='orange', width=1), name='MA10'))
    fig.add_trace(go.Scatter(x=recent_df['Date'], y=recent_df['MA50'], line=dict(color='purple', width=1), name='MA50'))

    fig.update_layout(
        title=f'{ticker} - Candlestick Chart with Indicators (Last 120 Days)',
        yaxis_title='Stock Price (INR)',
        xaxis_title='Date',
        xaxis_rangeslider_visible=False,
        template='plotly_dark'
    )
    fig.show()

# --- Main function ---
def main():
    # 1. Get User Input
    while True:
        COMPANY_NAME = input("▶️ Enter company name (e.g., Tata Motors): ")
        TICKER = input(f"▶️ Enter Yahoo Finance ticker for '{COMPANY_NAME}' (e.g., TATAMOTORS.NS): ")
        print(f"\nValidating ticker '{TICKER}'...")
        stock_df_raw = fetch_stock_data(TICKER)
        if stock_df_raw is not None:
            print(f"✅ Ticker '{TICKER}' is valid."); break
        else:
            print(f"❌ Error: Ticker '{TICKER}' is invalid. Please try again.\n")

    # 2. Fetch Data
    print("\n--- Fetching & Processing Data ---")
    news_df = fetch_news(COMPANY_NAME)
    financial_events_df = fetch_real_events(TICKER, stock_df_raw)
    news_events_df = find_events_from_news(news_df)
    events_df = pd.concat([financial_events_df, news_events_df], ignore_index=True)

    # 3. Sentiment & Feature Engineering
    if not news_df.empty:
        try:
            tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
            model_fbert = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone").to(DEVICE)
            def get_sentiment(text):
                if not text or not isinstance(text, str): return 0.0
                inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True).to(DEVICE)
                with torch.no_grad(): outputs = model_fbert(**inputs)
                scores = torch.nn.functional.softmax(outputs.logits, dim=1).cpu().numpy()[0]; return scores[1] - scores[0]
            tqdm.pandas(desc="Analyzing headlines"); news_df['Sentiment'] = news_df['Headline'].progress_apply(get_sentiment)
            sentiment_df = news_df.groupby("Date")["Sentiment"].mean().reset_index(); sentiment_df['Date'] = pd.to_datetime(sentiment_df['Date'])
        except Exception as e:
            sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])
    else:
        sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])

    stock_df = add_features(stock_df_raw)
    if not events_df.empty:
        event_dummies = pd.get_dummies(events_df['Event_Type'], prefix='Event', dtype=float)
        events_df = pd.concat([events_df['Date'], event_dummies], axis=1).groupby('Date').sum().reset_index()
        stock_df = pd.merge(stock_df, events_df, on='Date', how='left')
    stock_df = pd.merge(stock_df, sentiment_df, on='Date', how='left') if not sentiment_df.empty else stock_df.assign(Sentiment=0.0)

    event_cols = [col for col in stock_df.columns if 'Event_' in col]
    stock_df['Target'] = stock_df['Return'].shift(-1)
    stock_df[event_cols] = stock_df[event_cols].fillna(0); stock_df['Sentiment'].fillna(0, inplace=True)
    # MODIFIED: Added MACD_Signal to the list of columns to fill
    tech_cols = ['Return', 'MA10', 'MA50', 'RSI', 'MACD', 'MACD_Signal', 'Volatility']
    stock_df[tech_cols] = stock_df[tech_cols].fillna(method='ffill').fillna(method='bfill')
    stock_df.replace([np.inf, -np.inf], 0, inplace=True); stock_df.dropna(inplace=True)

    # 4. Normalize & Split Data
    features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA10', 'MA50', 'RSI', 'MACD', 'MACD_Signal', 'Return', 'Volatility', 'Sentiment'] + event_cols
    scaler = MinMaxScaler(); stock_df_scaled = stock_df.copy()
    stock_df_scaled[features] = scaler.fit_transform(stock_df_scaled[features])
    train_size = int(len(stock_df_scaled) * 0.85)
    train_df, test_df = stock_df_scaled.iloc[:train_size].copy(), stock_df_scaled.iloc[train_size:].copy()
    test_df_unscaled = stock_df.iloc[train_size:].copy()

    # 5. Create Datasets and Dataloaders
    train_dataset = StockDataset(train_df, train_df, features, 'Target', SEQ_LEN)
    test_dataset = StockDataset(test_df, test_df_unscaled, features, 'Target', SEQ_LEN)
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

    # 6. Train Models
    input_size = len(features); hidden_size = 64
    num_heads = 4 if input_size % 4 == 0 else (2 if input_size % 2 == 0 else 1)
    print(f"\nTotal input features: {input_size}. Using {num_heads} attention heads.")
    tlstm_model = train_model(TLSTM_Attention(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
    gru_model = train_model(GRU_Model(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
    trans_model = train_model(TransformerBlock(input_size, num_heads, 64).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)

    # 7. Evaluate
    tlstm_preds = get_predictions(tlstm_model, test_loader)
    gru_preds = get_predictions(gru_model, test_loader)
    trans_preds = get_predictions(trans_model, test_loader)
    final_predicted_returns = (tlstm_preds + gru_preds + trans_preds) / 3.0
    predicted_prices = test_dataset.close_prices_for_eval * (1 + final_predicted_returns)
    actual_prices = test_df_unscaled['Close'].values[SEQ_LEN:]

    print("\n--- Final Ensemble Model Evaluation ---")
    mae = mean_absolute_error(actual_prices, predicted_prices)
    mse = mean_squared_error(actual_prices, predicted_prices)
    r2 = r2_score(actual_prices, predicted_prices)
    print(f"Mean Absolute Error (MAE): ₹{mae:.2f}")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"R-squared (R²): {r2:.4f}")

    # 8. Plot Line Chart of Predictions
    plt.style.use('seaborn-v0_8-darkgrid')
    plt.figure(figsize=(15, 7))
    plt.plot(test_df_unscaled['Date'][SEQ_LEN:], actual_prices, label='Actual Price', color='royalblue')
    plt.plot(test_df_unscaled['Date'][SEQ_LEN:], predicted_prices, label='Predicted Price', color='crimson', linestyle='--')
    plt.title(f"{TICKER} Stock Price Prediction", fontsize=16); plt.xlabel("Date"); plt.ylabel("Close Price (INR)"); plt.legend(); plt.show()

    # --- 9. NEW: Generate Signal and Candlestick Chart ---
    last_day_unscaled_data = stock_df.iloc[-1]
    predicted_next_day_return = final_predicted_returns[-1]

    generate_trading_signal(last_day_unscaled_data, predicted_next_day_return)
    plot_candlestick_with_signals(stock_df, TICKER)

if __name__ == "__main__":
    main()
```

```python
def plot_event_impact(df, event_cols):
    """
    Analyzes and plots the average price return on days with specific events.
    """
    df['Price_Change'] = df['Close'].pct_change() * 100

    event_impacts = {}
    for col in event_cols:
        # Get the mean price change on days where this event occurred
        impact = df[df[col] == 1]['Price_Change'].mean()
        event_impacts[col.replace('Event_', '')] = impact

    # Calculate the average price change on a normal day (no events)
    all_event_days = df[df[event_cols].sum(axis=1) > 0].index
    non_event_mean = df.drop(all_event_days)['Price_Change'].mean()
    event_impacts['Non-Event Day'] = non_event_mean

    impact_df = pd.DataFrame(list(event_impacts.items()), columns=['Event', 'Avg_Return_Pct']).sort_values(by='Avg_Return_Pct')

    plt.figure(figsize=(10, 6))
    colors = ['crimson' if x < 0 else 'limegreen' for x in impact_df['Avg_Return_Pct']]
    sns.barplot(x='Avg_Return_Pct', y='Event', data=impact_df, palette=colors, orient='h')
    plt.title('Average Price Return (%) on Event Days vs. Non-Event Days', fontsize=16)
    plt.xlabel('Average Daily Return (%)')
    plt.ylabel('Event Type')
    plt.axvline(x=0, color='grey', linestyle='--')
    plt.tight_layout()
    plt.show()
```

```python
# 1. INSTALL ALL NECESSARY LIBRARIES
!pip install yfinance transformers torch scikit-learn requests tqdm plotly seaborn

# 2. IMPORT LIBRARIES
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from datetime import datetime, timedelta
from tqdm import tqdm
import requests
import yfinance as yf
import warnings
import re

warnings.filterwarnings('ignore')

# 3. SETUP CONSTANTS
SEQ_LEN = 30
BATCH_SIZE = 32
EPOCHS = 100
LEARNING_RATE = 0.0001
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
# --- IMPORTANT: Replace with your own NewsAPI key ---
NEWS_API_KEY = 'YOUR_NEWS_API_KEY' # <--- REPLACE THIS

print(f"Using device: {DEVICE}")
if NEWS_API_KEY == 'YOUR_NEWS_API_KEY':
    print("\nWARNING: Please replace 'YOUR_NEWS_API_KEY' with your actual NewsAPI key to fetch news data.")


# 4. DATA FETCHING AND PROCESSING FUNCTIONS
def fetch_stock_data(ticker):
    """Fetches daily stock data for the last 5 years and cleans it."""
    df = yf.download(tickers=ticker, period="5y", interval="1d", progress=False)
    if df.empty: return None
    if isinstance(df.columns, pd.MultiIndex): df.columns = df.columns.get_level_values(0)
    df.reset_index(inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    return df

def fetch_news(symbol):
    """Fetches news from the last 30 days from NewsAPI."""
    if NEWS_API_KEY == 'YOUR_NEWS_API_KEY':
        return pd.DataFrame(columns=["Date", "Headline"])
    to_date = datetime.today().strftime('%Y-%m-%d')
    from_date = (datetime.today() - timedelta(days=30)).strftime('%Y-%m-%d')
    url = f"https://newsapi.org/v2/everything?q={symbol}&from={from_date}&to={to_date}&language=en&sortBy=publishedAt&apiKey={NEWS_API_KEY}"
    try:
        res = requests.get(url).json()
        if res.get("status") == "ok":
            articles = res.get("articles", [])
            headlines = [{"Date": a['publishedAt'][:10], "Headline": a['title']} for a in articles]
            news_df = pd.DataFrame(headlines)
            news_df['Date'] = pd.to_datetime(news_df['Date'])
            return news_df
        else:
            return pd.DataFrame(columns=["Date", "Headline"])
    except Exception:
        return pd.DataFrame(columns=["Date", "Headline"])

def fetch_real_events(ticker_symbol, stock_df):
    """Fetches real financial events from yfinance and includes general market events."""
    all_events, ticker_obj = [], yf.Ticker(ticker_symbol)
    try: # Earnings
        earnings_dates = ticker_obj.get_earnings_dates(limit=20)
        if earnings_dates is not None and not earnings_dates.empty:
            all_events.append(pd.DataFrame({'Date': pd.to_datetime(earnings_dates.index.date), 'Event_Type': 'Earnings'}))
    except Exception: pass
    try: # Dividends & Splits
        actions_df = ticker_obj.actions
        if actions_df is not None and not actions_df.empty:
            actions_df.reset_index(inplace=True)
            actions_df['Date'] = pd.to_datetime(actions_df['Date'].dt.date)
            dividends = actions_df[actions_df['Dividends'] > 0][['Date']].copy(); dividends['Event_Type'] = 'Dividend'
            splits = actions_df[actions_df['Stock Splits'] > 0][['Date']].copy(); splits['Event_Type'] = 'Stock Split'
            if not dividends.empty: all_events.append(dividends)
            if not splits.empty: all_events.append(splits)
    except Exception: pass
    rbi_mpc_dates = ['2020-02-06', '2020-03-27', '2020-05-22', '2020-08-06', '2020-10-09', '2020-12-04', '2021-02-05', '2021-04-07', '2021-06-04', '2021-08-06', '2021-10-08', '2021-12-08', '2022-02-10', '2022-04-08', '2022-05-04', '2022-06-08', '2022-08-05', '2022-09-30', '2022-12-07', '2023-02-08', '2023-04-06', '2023-06-08', '2023-08-10', '2023-10-06', '2023-12-08', '2024-02-08', '2024-04-05', '2024-06-07', '2024-08-08', '2024-10-09', '2024-12-06', '2025-02-07', '2025-04-04', '2025-06-06', '2025-08-08']
    all_events.append(pd.DataFrame({'Date': pd.to_datetime(rbi_mpc_dates), 'Event_Type': 'RBI MPC Meeting'}))
    if not all_events: return pd.DataFrame(columns=['Date'])
    events_df = pd.concat(all_events, ignore_index=True)
    min_date, max_date = stock_df['Date'].min(), stock_df['Date'].max()
    events_df = events_df[(events_df['Date'] >= min_date) & (events_df['Date'] <= max_date)]
    return events_df.drop_duplicates(subset=['Date', 'Event_Type']).reset_index(drop=True)

def find_events_from_news(news_df):
    """Scans news headlines for keywords to identify major announcements."""
    keyword_map = {'Major Announcement': [r'AGM', r'annual general meeting'], 'Product Launch': [r'launch', r'unveil'], 'Merger/Acquisition': [r'merger', r'acquire', r'takeover'], 'Partnership/Deal': [r'partnership', r'deal']}
    found_events = []
    if news_df.empty: return pd.DataFrame(columns=['Date', 'Event_Type'])
    for _, row in news_df.iterrows():
        headline = row['Headline'].lower()
        for event_type, keywords in keyword_map.items():
            if any(re.search(keyword, headline) for keyword in keywords):
                found_events.append({'Date': row['Date'], 'Event_Type': event_type}); break
    if not found_events: return pd.DataFrame(columns=['Date', 'Event_Type'])
    return pd.DataFrame(found_events).drop_duplicates()

def add_features(df):
    """Calculates technical indicators and adds them as features."""
    df['Return'] = df['Close'].pct_change()
    df['MA10'] = df['Close'].rolling(window=10).mean()
    df['MA50'] = df['Close'].rolling(window=50).mean()
    delta = df['Close'].diff()
    up, down = delta.clip(lower=0), -1*delta.clip(upper=0)
    ema_up, ema_down = up.ewm(com=13, adjust=False).mean(), down.ewm(com=13, adjust=False).mean()
    rs = ema_up / ema_down
    df['RSI'] = 100 - (100 / (1 + rs))
    df['MACD'] = df['Close'].ewm(span=12, adjust=False).mean() - df['Close'].ewm(span=26, adjust=False).mean()
    df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()
    df['Volatility'] = df['Return'].rolling(window=21).std() * np.sqrt(252)
    return df


# 5. MODEL DEFINITIONS
class TLSTM_Attention(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(TLSTM_Attention, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.attention = nn.Linear(hidden_size, 1); self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        lstm_out, _ = self.lstm(x); attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
        context = torch.sum(attn_weights * lstm_out, dim=1); return self.fc(context)

class GRU_Model(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(GRU_Model, self).__init__(); self.gru = nn.GRU(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.gru(x); return self.fc(out[:, -1, :])

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.attn = nn.MultiheadAttention(embed_dim, num_heads, batch_first=True, dropout=rate)
        self.ffn = nn.Sequential(nn.Linear(embed_dim, ff_dim), nn.ReLU(), nn.Linear(ff_dim, embed_dim))
        self.layernorm1, self.layernorm2 = nn.LayerNorm(embed_dim, eps=1e-6), nn.LayerNorm(embed_dim, eps=1e-6)
        self.dropout1, self.dropout2 = nn.Dropout(rate), nn.Dropout(rate); self.fc_out = nn.Linear(embed_dim, 1)
    def forward(self, x):
        attn_output, _ = self.attn(x, x, x); out1 = self.layernorm1(x + self.dropout1(attn_output))
        ffn_output = self.ffn(out1); out2 = self.layernorm2(out1 + self.dropout2(ffn_output))
        return self.fc_out(out2.mean(dim=1))


# 6. TRAINING, DATASET, AND EVALUATION HELPERS
class StockDataset(Dataset):
    def __init__(self, df, df_unscaled, features, target_col, seq_len):
        self.X, self.y, self.close_prices_for_eval = [], [], df_unscaled['Close'].values[seq_len-1:-1]
        feature_data, target_data = df[features].values, df[target_col].values
        for i in range(seq_len, len(df)): self.X.append(feature_data[i-seq_len:i]); self.y.append(target_data[i])
        self.X, self.y = np.array(self.X), np.array(self.y)
    def __len__(self): return len(self.X)
    def __getitem__(self, idx): return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)

def train_model(model, train_loader, epochs, lr):
    loss_fn = nn.MSELoss(); optimizer = optim.Adam(model.parameters(), lr=lr); model.train()
    print(f"\n--- Training {model.__class__.__name__} ---")
    for epoch in range(epochs):
        total_loss = 0; pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")
        for X_batch, y_batch in pbar:
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE); optimizer.zero_grad()
            preds = model(X_batch).squeeze(-1); loss = loss_fn(preds, y_batch); loss.backward(); optimizer.step()
            total_loss += loss.item(); pbar.set_postfix({'Loss': f"{total_loss / len(pbar):.8f}"})
    return model

def get_predictions(model, loader):
    model.eval(); preds = []
    with torch.no_grad():
        for X_batch, _ in loader: preds.append(model(X_batch.to(DEVICE)).squeeze(-1).cpu().numpy())
    return np.concatenate(preds)


# 7. VISUALIZATION AND SIGNALING FUNCTIONS
def generate_trading_signal(last_day_data, predicted_next_return):
    """Generates a buy/don't buy signal based on a set of rules."""
    signal_checks = {
        "Model Prediction is Positive": predicted_next_return > 0,
        "Uptrend (MA10 > MA50)": last_day_data['MA10'] > last_day_data['MA50'],
        "Momentum is Bullish (MACD > Signal)": last_day_data['MACD'] > last_day_data['MACD_Signal'],
        "Not Overbought (RSI < 75)": last_day_data['RSI'] < 75
    }
    print("\n--- Trading Signal Analysis ---")
    print("Disclaimer: This is not financial advice. For educational purposes only.")
    print(f"Analysis for date: {last_day_data['Date'].strftime('%Y-%m-%d')}")
    print("-" * 30)
    all_conditions_met = True
    for check, result in signal_checks.items():
        print(f"▶️ {check}: {'✅' if result else '❌'}")
        if not result: all_conditions_met = False
    print("-" * 30)
    if all_conditions_met:
        print("Signal: 🟢 BUY 🟢")
        print("Reasoning: All bullish conditions were met, including a positive model prediction.")
    else:
        print("Signal: 🔴 HOLD / DO NOT BUY 🔴")
        print("Reasoning: One or more conditions for a buy signal were not met.")

def plot_candlestick_with_signals(df, ticker):
    """Plots an interactive candlestick chart for the last 120 days."""
    recent_df = df.tail(120)
    fig = go.Figure(data=[
        go.Candlestick(x=recent_df['Date'], open=recent_df['Open'], high=recent_df['High'], low=recent_df['Low'], close=recent_df['Close'], name='Price'),
        go.Scatter(x=recent_df['Date'], y=recent_df['MA10'], line=dict(color='orange', width=1), name='MA10'),
        go.Scatter(x=recent_df['Date'], y=recent_df['MA50'], line=dict(color='purple', width=1), name='MA50')
    ])
    fig.update_layout(title=f'{ticker} - Candlestick Chart (Last 120 Days)', yaxis_title='Stock Price (INR)', xaxis_rangeslider_visible=False, template='plotly_dark')
    fig.show()

def plot_feature_importance(df, features, target_col):
    """Trains a RandomForest model to estimate and plot feature importance."""
    X = df[features]; y = df[target_col]
    features_for_model = [f for f in features if f != 'MACD_Signal']
    X = X[features_for_model]
    model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1); model.fit(X, y)
    importance_df = pd.DataFrame({'Feature': features_for_model, 'Importance': model.feature_importances_}).sort_values(by='Importance', ascending=False)
    plt.figure(figsize=(10, 8)); sns.barplot(x='Importance', y='Feature', data=importance_df, palette='viridis')
    plt.title('Feature Importance for Stock Price Prediction', fontsize=16); plt.tight_layout(); plt.show()

def plot_correlation_heatmap(df, ticker):
    """Plots a heatmap of the correlation matrix for the main features."""
    corr_features = ['Close', 'Volume', 'Return', 'MA10', 'MA50', 'RSI', 'MACD', 'Volatility', 'Sentiment', 'Target']
    corr_matrix = df[corr_features].corr()
    plt.figure(figsize=(12, 10)); sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title(f'{ticker} - Feature Correlation Heatmap', fontsize=16); plt.show()

def plot_event_impact(df, event_cols):
    """Analyzes and plots the average price return on days with specific events."""
    df['Price_Change'] = df['Close'].pct_change() * 100; event_impacts = {}
    for col in event_cols:
        impact = df[df[col] == 1]['Price_Change'].mean()
        event_impacts[col.replace('Event_', '')] = impact
    all_event_days = df[df[event_cols].sum(axis=1) > 0].index
    non_event_mean = df.drop(all_event_days)['Price_Change'].mean()
    event_impacts['Non-Event Day'] = non_event_mean
    impact_df = pd.DataFrame(list(event_impacts.items()), columns=['Event', 'Avg_Return_Pct']).sort_values(by='Avg_Return_Pct')
    plt.figure(figsize=(10, 6)); colors = ['crimson' if x < 0 else 'limegreen' for x in impact_df['Avg_Return_Pct']]
    sns.barplot(x='Avg_Return_Pct', y='Event', data=impact_df, palette=colors, orient='h')
    plt.title('Average Price Return (%) on Event Days', fontsize=16); plt.xlabel('Average Daily Return (%)'); plt.axvline(x=0, color='grey', linestyle='--'); plt.tight_layout(); plt.show()

def plot_error_distribution(actual_prices, predicted_prices):
    """Plots the distribution of the prediction errors."""
    errors = actual_prices - predicted_prices
    plt.figure(figsize=(10, 6)); sns.histplot(errors, kde=True, bins=50, color='skyblue')
    plt.title('Distribution of Prediction Errors (Actual - Predicted)', fontsize=16); plt.xlabel('Prediction Error (in INR)'); plt.axvline(x=0, color='red', linestyle='--', label='Zero Error'); plt.legend(); plt.show()


# 8. MAIN EXECUTION FUNCTION
def main():
    # --- Get User Input ---
    while True:
        COMPANY_NAME = input("▶️ Enter company name (e.g., Tata Motors): ")
        TICKER = input(f"▶️ Enter Yahoo Finance ticker for '{COMPANY_NAME}' (e.g., TATAMOTORS.NS): ")
        print(f"\nValidating ticker '{TICKER}'...")
        stock_df_raw = fetch_stock_data(TICKER)
        if stock_df_raw is not None:
            print(f"✅ Ticker '{TICKER}' is valid."); break
        else:
            print(f"❌ Error: Ticker '{TICKER}' is invalid. Please try again.\n")

    # --- Fetch & Process Data ---
    print("\n--- Fetching & Processing Data ---")
    news_df = fetch_news(COMPANY_NAME)
    financial_events_df = fetch_real_events(TICKER, stock_df_raw)
    news_events_df = find_events_from_news(news_df)
    events_df = pd.concat([financial_events_df, news_events_df], ignore_index=True)

    # --- Sentiment & Feature Engineering ---
    if not news_df.empty:
        try:
            tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
            model_fbert = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone").to(DEVICE)
            def get_sentiment(text):
                if not text or not isinstance(text, str): return 0.0
                inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True).to(DEVICE)
                with torch.no_grad(): outputs = model_fbert(**inputs)
                scores = torch.nn.functional.softmax(outputs.logits, dim=1).cpu().numpy()[0]; return scores[1] - scores[0]
            tqdm.pandas(desc="Analyzing headlines"); news_df['Sentiment'] = news_df['Headline'].progress_apply(get_sentiment)
            sentiment_df = news_df.groupby("Date")["Sentiment"].mean().reset_index(); sentiment_df['Date'] = pd.to_datetime(sentiment_df['Date'])
        except Exception: sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])
    else: sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])

    stock_df = add_features(stock_df_raw)
    if not events_df.empty:
        event_dummies = pd.get_dummies(events_df['Event_Type'], prefix='Event', dtype=float)
        events_df = pd.concat([events_df['Date'], event_dummies], axis=1).groupby('Date').sum().reset_index()
        stock_df = pd.merge(stock_df, events_df, on='Date', how='left')
    stock_df = pd.merge(stock_df, sentiment_df, on='Date', how='left') if not sentiment_df.empty else stock_df.assign(Sentiment=0.0)
    event_cols = [col for col in stock_df.columns if 'Event_' in col]
    stock_df['Target'] = stock_df['Return'].shift(-1)
    stock_df[event_cols] = stock_df[event_cols].fillna(0); stock_df['Sentiment'].fillna(0, inplace=True)
    tech_cols = ['Return', 'MA10', 'MA50', 'RSI', 'MACD', 'MACD_Signal', 'Volatility']
    stock_df[tech_cols] = stock_df[tech_cols].fillna(method='ffill').fillna(method='bfill')
    stock_df.replace([np.inf, -np.inf], 0, inplace=True); stock_df.dropna(inplace=True)

    # --- Normalize & Split Data ---
    features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA10', 'MA50', 'RSI', 'MACD', 'MACD_Signal', 'Return', 'Volatility', 'Sentiment'] + event_cols
    scaler = MinMaxScaler(); stock_df_scaled = stock_df.copy()
    stock_df_scaled[features] = scaler.fit_transform(stock_df_scaled[features])
    train_size = int(len(stock_df_scaled) * 0.85)
    train_df, test_df = stock_df_scaled.iloc[:train_size].copy(), stock_df_scaled.iloc[train_size:].copy()
    test_df_unscaled = stock_df.iloc[train_size:].copy()

    # --- Create Datasets & Dataloaders ---
    train_dataset = StockDataset(train_df, train_df, features, 'Target', SEQ_LEN)
    test_dataset = StockDataset(test_df, test_df_unscaled, features, 'Target', SEQ_LEN)
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

    # --- Train Models ---
    input_size = len(features); hidden_size = 64
    num_heads = 4 if input_size % 4 == 0 else (2 if input_size % 2 == 0 else 1)
    print(f"\nTotal input features: {input_size}. Using {num_heads} attention heads for Transformer.")
    tlstm_model = train_model(TLSTM_Attention(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
    gru_model = train_model(GRU_Model(input_size, hidden_size).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)
    trans_model = train_model(TransformerBlock(input_size, num_heads, 64).to(DEVICE), train_loader, EPOCHS, LEARNING_RATE)

    # --- Evaluate Ensemble Model ---
    tlstm_preds, gru_preds, trans_preds = get_predictions(tlstm_model, test_loader), get_predictions(gru_model, test_loader), get_predictions(trans_model, test_loader)
    final_predicted_returns = (tlstm_preds + gru_preds + trans_preds) / 3.0
    predicted_prices = test_dataset.close_prices_for_eval * (1 + final_predicted_returns)
    actual_prices = test_df_unscaled['Close'].values[SEQ_LEN:]

    print("\n--- Final Ensemble Model Evaluation ---")
    print(f"Mean Absolute Error (MAE): ₹{mean_absolute_error(actual_prices, predicted_prices):.2f}")
    print(f"Mean Squared Error (MSE): {mean_squared_error(actual_prices, predicted_prices):.2f}")
    print(f"R-squared (R²): {r2_score(actual_prices, predicted_prices):.4f}")

    # --- Generate Trading Signal for Next Day ---
    last_day_unscaled_data = stock_df.iloc[-1]
    predicted_next_day_return = final_predicted_returns[-1]
    generate_trading_signal(last_day_unscaled_data, predicted_next_day_return)

    # --- Generate All Visualizations ---
    print("\n--- Generating Visualizations ---")
    plt.style.use('seaborn-v0_8-darkgrid')

    # 1. Prediction Plot
    plt.figure(figsize=(15, 7)); plt.plot(test_df_unscaled['Date'][SEQ_LEN:], actual_prices, label='Actual Price', color='royalblue')
    plt.plot(test_df_unscaled['Date'][SEQ_LEN:], predicted_prices, label='Predicted Price', color='crimson', linestyle='--')
    plt.title(f"{TICKER} Stock Price Prediction", fontsize=16); plt.xlabel("Date"); plt.ylabel("Close Price (INR)"); plt.legend(); plt.show()

    # 2. Candlestick Chart
    plot_candlestick_with_signals(stock_df, TICKER)

    # 3. Advanced Visualizations
    plot_feature_importance(stock_df, features, 'Target')
    plot_correlation_heatmap(stock_df, TICKER)
    if event_cols: plot_event_impact(stock_df.copy(), event_cols)
    plot_error_distribution(actual_prices, predicted_prices)


# 9. RUN THE SCRIPT
if __name__ == "__main__":
    main()
```

```python
# NEW: Install necessary libraries, including for web scraping
!pip install yfinance transformers torch scikit-learn requests tqdm beautifulsoup4 lxml

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from datetime import datetime
from dateutil.relativedelta import relativedelta
from tqdm import tqdm
import requests
import yfinance as yf
import warnings
import copy
import re
from bs4 import BeautifulSoup

warnings.filterwarnings('ignore')

# --- Constants ---
SEQ_LEN = 30
BATCH_SIZE = 32
EPOCHS = 100
LEARNING_RATE = 0.0001
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
TRAIN_SPLIT = 0.70
VALIDATION_SPLIT = 0.15
PATIENCE = 10
# --- IMPORTANT: Replace with your own NewsAPI key ---
NEWS_API_KEY = '99d1937942dd430cb7e67bbd39c0a713' # <--- REPLACE THIS

print(f"Using device: {DEVICE}")

# --- DYNAMIC RBI DATA SCRAPER ---
def fetch_rbi_mpc_data_dynamically():
    """
    Dynamically scrapes the RBI website for MPC press releases to extract repo rate changes.

    WARNING: This is a web scraper and is highly dependent on the RBI website's
    HTML structure. It may break if the website is updated.
    """
    print("Attempting to dynamically scrape RBI for MPC decisions...")
    base_url = "https://www.rbi.org.in"
    press_releases_url = f"{base_url}/Scripts/BS_PressReleaseDisplay.aspx"

    try:
        # Fetch the main press releases page
        headers = {'User-Agent': 'Mozilla/5.0'}
        response = requests.get(press_releases_url, headers=headers)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'lxml')

        # Find links that contain "Monetary Policy Statement" or "MPC Resolution"
        mpc_links = soup.find_all('a', href=True, string=re.compile(r'Monetary Policy Statement|MPC Resolution|Monetary Policy Report'))

        if not mpc_links:
            print("Warning: Could not find any MPC links on the RBI press release page. Using an empty dataset.")
            return {}

        rbi_data = {}
        # Limit to the last 25 releases to be efficient
        for link in tqdm(mpc_links[:25], desc="Parsing RBI Releases"):
            release_url = f"{base_url}{link['href']}" if link['href'].startswith('/Scripts') else link['href']

            # Fetch the individual release page
            release_response = requests.get(release_url, headers=headers)
            release_soup = BeautifulSoup(release_response.content, 'lxml')

            # Extract text and date
            full_text = release_soup.get_text()
            date_match = re.search(r'(January|February|March|April|May|June|July|August|September|October|November|December)\s(\d{1,2}),\s(\d{4})', full_text)
            if not date_match: continue

            date_str = date_match.group(0)
            decision_date = pd.to_datetime(date_str).strftime('%Y-%m-%d')

            # Use regex to find the decision
            rate_change = 0.0

            # Search for increase
            increase_match = re.search(r'increase the policy repo rate.*?by\s+(\d+)\s+basis points', full_text, re.IGNORECASE)
            if increase_match:
                rate_change = float(increase_match.group(1)) / 100.0
            else:
                # Search for decrease
                decrease_match = re.search(r'reduce the policy repo rate.*?by\s+(\d+)\s+basis points', full_text, re.IGNORECASE)
                if decrease_match:
                    rate_change = -float(decrease_match.group(1)) / 100.0

            # If the decision date is not already captured, add it. This avoids duplicates.
            if decision_date not in rbi_data:
                 rbi_data[decision_date] = rate_change

        print(f"Successfully scraped {len(rbi_data)} unique MPC decisions.")
        return rbi_data

    except Exception as e:
        print(f"Error scraping RBI website: {e}. The scraper may need to be updated. Proceeding without RBI data.")
        return {}


# --- Data Fetching & Cleaning (Unchanged from before) ---
def fetch_stock_data(ticker):
    df = yf.download(tickers=ticker, period="5y", interval="1d", progress=False)
    if df.empty: return None
    df.columns = df.columns.get_level_values(0) if isinstance(df.columns, pd.MultiIndex) else df.columns
    df.reset_index(inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    return df

def fetch_historical_news(symbol, start_date, end_date):
    if NEWS_API_KEY == 'YOUR_NEWS_API_KEY':
        print("Warning: NewsAPI key not provided. Skipping news fetching.")
        return pd.DataFrame(columns=["Date", "Headline"])

    print(f"Fetching historical news from {start_date.date()} to {end_date.date()}.")
    all_headlines = []
    current_date = end_date
    while current_date > start_date:
        from_date = (current_date - relativedelta(months=1)).strftime('%Y-%m-%d')
        to_date = current_date.strftime('%Y-%m-%d')
        url = f"https://newsapi.org/v2/everything?q={symbol}&from={from_date}&to={to_date}&language=en&sortBy=publishedAt&apiKey={NEWS_API_KEY}"
        try:
            res = requests.get(url).json()
            if res.get("status") == "ok":
                articles = res.get("articles", [])
                headlines = [{"Date": a['publishedAt'][:10], "Headline": a['title']} for a in articles]
                all_headlines.extend(headlines)
            else:
                print(f"Stopping news fetch at {from_date}: {res.get('message')}")
                break
        except Exception as e:
            print(f"An exception occurred during news fetching: {e}")
            break
        current_date -= relativedelta(months=1)

    return pd.DataFrame(all_headlines).drop_duplicates() if all_headlines else pd.DataFrame(columns=["Date", "Headline"])

def fetch_real_events(ticker_symbol, stock_df):
    """
    FULLY DYNAMIC: Fetches real financial events.
    - Earnings Surprise % from yfinance (if available).
    - RBI MPC decisions are now scraped dynamically.
    """
    all_events = []
    min_date, max_date = stock_df['Date'].min(), stock_df['Date'].max()

    # 1. Fetch Earnings Data (already dynamic)
    try:
        ticker_obj = yf.Ticker(ticker_symbol)
        earnings_dates = ticker_obj.get_earnings_dates(limit=20)
        if earnings_dates is not None and not earnings_dates.empty and 'EPS Surprise(%)' in earnings_dates.columns:
            print("Found 'EPS Surprise(%)' data.")
            earnings_df = earnings_dates.reset_index()[['Earnings Date', 'EPS Surprise(%)']]
            earnings_df.columns = ['Date', 'Earnings_Surprise_Pct']
            earnings_df['Date'] = pd.to_datetime(earnings_df['Date'].dt.date)
            earnings_df['Earnings_Surprise_Pct'] = earnings_df['Earnings_Surprise_Pct'] / 100.0
            all_events.append(earnings_df)
        else:
            print("Warning: 'EPS Surprise(%)' data not found for this ticker. Skipping this feature.")
    except Exception as e:
        print(f"Could not fetch earnings dates: {e}")

    # 2. DYNAMICALLY fetch RBI MPC Meeting Data
    rbi_mpc_data = fetch_rbi_mpc_data_dynamically()
    if rbi_mpc_data:
        rbi_df = pd.DataFrame(list(rbi_mpc_data.items()), columns=['Date', 'RBI_Rate_Change'])
        rbi_df['Date'] = pd.to_datetime(rbi_df['Date'])
        all_events.append(rbi_df)

    if not all_events: return pd.DataFrame()

    events_df = pd.concat(all_events, ignore_index=True).groupby('Date').sum().reset_index()
    return events_df[(events_df['Date'] >= min_date) & (events_df['Date'] <= max_date)].reset_index(drop=True)


# --- All other functions (add_features, Models, train_model, etc.) remain exactly the same ---
def add_features(df):
    df['Return'] = df['Close'].pct_change()
    df['MA10'] = df['Close'].rolling(window=10).mean()
    df['MA50'] = df['Close'].rolling(window=50).mean()
    delta = df['Close'].diff()
    up, down = delta.clip(lower=0), -1*delta.clip(upper=0)
    ema_up, ema_down = up.ewm(com=13, adjust=False).mean(), down.ewm(com=13, adjust=False).mean()
    rs = ema_up / ema_down
    df['RSI'] = 100 - (100 / (1 + rs))
    df['MACD'] = df['Close'].ewm(span=12, adjust=False).mean() - df['Close'].ewm(span=26, adjust=False).mean()
    df['Volatility'] = df['Return'].rolling(window=21).std() * np.sqrt(252)
    return df

class TLSTM_Attention(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(TLSTM_Attention, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.attention = nn.Linear(hidden_size, 1)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        attn_weights = torch.softmax(self.attention(lstm_out), dim=1)
        context = torch.sum(attn_weights * lstm_out, dim=1)
        return self.fc(context)

class GRU_Model(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers=2):
        super(GRU_Model, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=0.2)
        self.fc = nn.Linear(hidden_size, 1)
    def forward(self, x):
        out, _ = self.gru(x)
        return self.fc(out[:, -1, :])

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.attn = nn.MultiheadAttention(embed_dim, num_heads, batch_first=True, dropout=rate)
        self.ffn = nn.Sequential(nn.Linear(embed_dim, ff_dim), nn.ReLU(), nn.Linear(ff_dim, embed_dim))
        self.layernorm1, self.layernorm2 = nn.LayerNorm(embed_dim, eps=1e-6), nn.LayerNorm(embed_dim, eps=1e-6)
        self.dropout1, self.dropout2 = nn.Dropout(rate), nn.Dropout(rate)
        self.fc_out = nn.Linear(embed_dim, 1)
    def forward(self, x):
        attn_output, _ = self.attn(x, x, x)
        out1 = self.layernorm1(x + self.dropout1(attn_output))
        ffn_output = self.ffn(out1)
        out2 = self.layernorm2(out1 + self.dropout2(ffn_output))
        return self.fc_out(out2.mean(dim=1))

def train_model(model, train_loader, val_loader, epochs, lr, patience):
    loss_fn = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    print(f"\n--- Training {model.__class__.__name__} ---")
    best_val_loss = float('inf')
    epochs_no_improve = 0
    best_model_state = None
    for epoch in range(epochs):
        model.train()
        total_train_loss = 0
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs} [T]", leave=False)
        for X_batch, y_batch in pbar:
            X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
            optimizer.zero_grad()
            preds = model(X_batch).squeeze(-1)
            loss = loss_fn(preds, y_batch)
            loss.backward()
            optimizer.step()
            total_train_loss += loss.item()
            pbar.set_postfix({'Train Loss': f"{total_train_loss / len(pbar):.6f}"})
        model.eval()
        total_val_loss = 0
        with torch.no_grad():
            for X_batch, y_batch in val_loader:
                X_batch, y_batch = X_batch.to(DEVICE), y_batch.to(DEVICE)
                preds = model(X_batch).squeeze(-1)
                val_loss = loss_fn(preds, y_batch)
                total_val_loss += val_loss.item()
        avg_val_loss = total_val_loss / len(val_loader)
        print(f"Epoch {epoch+1}/{epochs} -> Avg Train Loss: {total_train_loss / len(train_loader):.6f}, Avg Val Loss: {avg_val_loss:.6f}")
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            epochs_no_improve = 0
            best_model_state = copy.deepcopy(model.state_dict())
        else:
            epochs_no_improve += 1
        if epochs_no_improve >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs.")
            break
    if best_model_state: model.load_state_dict(best_model_state)
    return model, best_val_loss

def get_predictions(model, loader):
    model.eval()
    preds = []
    with torch.no_grad():
        for X_batch, _ in loader:
            preds.append(model(X_batch.to(DEVICE)).squeeze(-1).cpu().numpy())
    return np.concatenate(preds)

class StockDataset(Dataset):
    def __init__(self, df, df_unscaled, features, target_col, seq_len):
        self.X, self.y, self.close_prices_for_eval = [], [], df_unscaled['Close'].values[seq_len-1:-1]
        feature_data, target_data = df[features].values, df[target_col].values
        for i in range(seq_len, len(df)):
            self.X.append(feature_data[i-seq_len:i])
            self.y.append(target_data[i])
        self.X, self.y = np.array(self.X), np.array(self.y)
    def __len__(self): return len(self.X)
    def __getitem__(self, idx): return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)

def main():
    while True:
        COMPANY_NAME = input("▶️ Enter the full company name for news search (e.g., Tata Motors): ")
        TICKER = input(f"▶️ Enter the Yahoo Finance ticker for '{COMPANY_NAME}' (e.g., TATAMOTORS.NS): ")
        print(f"\nValidating ticker '{TICKER}'...")
        stock_df = fetch_stock_data(TICKER)
        if stock_df is not None:
            print(f"✅ Ticker '{TICKER}' is valid. Starting analysis...")
            break
        else:
            print(f"❌ Error: Ticker '{TICKER}' is invalid or no data found. Please try again.")

    print("\n--- Fetching Data ---")
    news_df = fetch_historical_news(COMPANY_NAME, stock_df['Date'].min(), stock_df['Date'].max())
    events_df = fetch_real_events(TICKER, stock_df)

    if not news_df.empty:
        print("\n--- Calculating News Sentiment ---")
        try:
            tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
            model_fbert = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone").to(DEVICE)
            def get_sentiment(text):
                if not text or not isinstance(text, str): return 0.0
                inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512, padding=True).to(DEVICE)
                with torch.no_grad(): outputs = model_fbert(**inputs)
                scores = torch.nn.functional.softmax(outputs.logits, dim=1).cpu().numpy()[0]
                return scores[1] - scores[0]
            tqdm.pandas(desc="Analyzing headlines")
            news_df['Sentiment'] = news_df['Headline'].progress_apply(get_sentiment)
            sentiment_df = news_df.groupby("Date")["Sentiment"].mean().reset_index()
            sentiment_df['Date'] = pd.to_datetime(sentiment_df['Date'])
        except Exception as e:
            print(f"Error during sentiment analysis: {e}")
            sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])
    else:
        sentiment_df = pd.DataFrame(columns=['Date', 'Sentiment'])

    print("\n--- Engineering & Merging Features ---")
    stock_df = add_features(stock_df)
    if not events_df.empty: stock_df = pd.merge(stock_df, events_df, on='Date', how='left')
    stock_df = pd.merge(stock_df, sentiment_df, on='Date', how='left')
    event_cols = ['Earnings_Surprise_Pct', 'RBI_Rate_Change']
    for col in event_cols:
        if col not in stock_df.columns: stock_df[col] = 0.0
    stock_df['Target'] = stock_df['Return'].shift(-1)
    stock_df[event_cols] = stock_df[event_cols].fillna(0)
    stock_df['Sentiment'].fillna(0, inplace=True)
    tech_cols = ['Return', 'MA10', 'MA50', 'RSI', 'MACD', 'Volatility']
    stock_df[tech_cols] = stock_df[tech_cols].fillna(method='ffill').fillna(method='bfill')
    stock_df.replace([np.inf, -np.inf], 0, inplace=True)
    stock_df.dropna(inplace=True)

    features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA10', 'MA50', 'RSI', 'MACD', 'Return', 'Volatility', 'Sentiment'] + event_cols
    print(f"Features being used: {features}")
    scaler = MinMaxScaler()
    stock_df_scaled = stock_df.copy()
    stock_df_scaled[features] = scaler.fit_transform(stock_df_scaled[features])

    train_size = int(len(stock_df_scaled) * TRAIN_SPLIT)
    val_size = int(len(stock_df_scaled) * VALIDATION_SPLIT)
    train_df = stock_df_scaled.iloc[:train_size]
    val_df = stock_df_scaled.iloc[train_size:train_size + val_size]
    test_df = stock_df_scaled.iloc[train_size + val_size:]
    test_df_unscaled = stock_df.iloc[train_size + val_size:]

    train_dataset = StockDataset(train_df, train_df, features, 'Target', SEQ_LEN)
    val_dataset = StockDataset(val_df, val_df, features, 'Target', SEQ_LEN)
    test_dataset = StockDataset(test_df, test_df_unscaled, features, 'Target', SEQ_LEN)

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

    input_size = len(features)
    hidden_size = 64
    num_heads = 4 if input_size % 4 == 0 else (2 if input_size % 2 == 0 else 1)
    print(f"\nTotal input features: {input_size}. Using {num_heads} attention heads for Transformer.")

    tlstm_model, v_loss_tlstm = train_model(TLSTM_Attention(input_size, hidden_size).to(DEVICE), train_loader, val_loader, EPOCHS, LEARNING_RATE, PATIENCE)
    gru_model, v_loss_gru = train_model(GRU_Model(input_size, hidden_size).to(DEVICE), train_loader, val_loader, EPOCHS, LEARNING_RATE, PATIENCE)
    trans_model, v_loss_trans = train_model(TransformerBlock(input_size, num_heads, 64).to(DEVICE), train_loader, val_loader, EPOCHS, LEARNING_RATE, PATIENCE)

    tlstm_preds = get_predictions(tlstm_model, test_loader)
    gru_preds = get_predictions(gru_model, test_loader)
    trans_preds = get_predictions(trans_model, test_loader)

    losses = [v_loss_tlstm, v_loss_gru, v_loss_trans]
    inv_losses = [1.0 / loss for loss in losses]
    total_inv_loss = sum(inv_losses)
    weights = [il / total_inv_loss for il in inv_losses]

    print("\n--- Weighted Ensemble Model ---")
    print(f"TLSTM Validation Loss: {v_loss_tlstm:.6f} -> Weight: {weights[0]:.3f}")
    print(f"GRU Validation Loss:   {v_loss_gru:.6f} -> Weight: {weights[1]:.3f}")
    print(f"Trans Validation Loss: {v_loss_trans:.6f} -> Weight: {weights[2]:.3f}")

    final_predicted_returns = (tlstm_preds * weights[0] + gru_preds * weights[1] + trans_preds * weights[2])
    last_actual_prices = test_dataset.close_prices_for_eval
    predicted_prices = last_actual_prices * (1 + final_predicted_returns)
    actual_prices = test_df_unscaled['Close'].values[SEQ_LEN:]

    print("\n--- Final Weighted Ensemble Model Evaluation ---")
    mae, mse, r2 = mean_absolute_error(actual_prices, predicted_prices), mean_squared_error(actual_prices, predicted_prices), r2_score(actual_prices, predicted_prices)
    print(f"Mean Absolute Error (MAE): ₹{mae:.2f}")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"R-squared (R²): {r2:.4f}")

    plt.style.use('seaborn-v0_8-darkgrid')
    plt.figure(figsize=(15, 7))
    plt.plot(test_df_unscaled['Date'].iloc[SEQ_LEN:], actual_prices, label='Actual Price', color='royalblue', linewidth=2)
    plt.plot(test_df_unscaled['Date'].iloc[SEQ_LEN:], predicted_prices, label='Predicted Price', color='crimson', linestyle='--')
    plt.title(f"{TICKER} Stock Price Prediction: Actual vs. Predicted (Weighted Ensemble)", fontsize=16)
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Close Price (INR)", fontsize=12)
    plt.legend()
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
